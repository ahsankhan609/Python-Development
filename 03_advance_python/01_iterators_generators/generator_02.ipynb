to make our **own iterator** we have to **write a long class**. it's much **tedious task** for the **developers**. so the Python provide the *generators* function. Means you don't need to write a long class. You can create a *generator* function.

*Generator* function gives the **iterator** object. and we don't need to make a **long class** for this.

python's generators are simple ways of creating iterators.

*Generator* is a function that returns an **iterator** object. You can take input values but not mandatory to pass.For example, in **decorator function** input was mandatory.

Generator object produce values one at a time. For example we need to generate 10 values, it will generate 1 value at a time.

Let's understands it with a example:

One family consumes 24 sacks of grains in one year. They have two options:

1. Take 24 sacks of grains and store at home
2. Take 2 sacks every month

which one is better?

If we choose **Option1**. The we need to take **24 sacks** of grains and **store** at home. It will **consume more space**.
So, the better option is **Option2**.

Syntax:-

```python
def generator_function():
    for i in range(10):
        yield i # due to yield keyword, it will be called as generator function.
# Yield keyword, only generate value, when it is called, otherwise no memory allocation will be done.
```



````python
from typing import TYPE_CHECKING

if TYPE_CHECKING:
    from collections.abc import Generator


def my_gen_func() -> Generator[int]:
    """Yield integers from 0 through 9, one value at a time."""
    yield from range(10)


gen_obj: Generator[int] = my_gen_func()
print(gen_obj)  # return memory address of the generator object.
print(type(gen_obj))  # <class 'generator'>

# next() fetches one value at a time; each call advances the generator.
print(next(gen_obj))  # 0
print(next(gen_obj))  # 1

# list() drains the rest of the generator in one go.
print(list(gen_obj))  # [2, 3, 4, 5, 6, 7, 8, 9]

# After a generator is exhausted, next() raises StopIteration.
try:
    next(gen_obj)
except StopIteration:
    print("StopIteration — generator has no more values")
````

````python
# another example with for loop

from typing import TYPE_CHECKING

if TYPE_CHECKING:
    from collections.abc import Generator


def my_gen() -> Generator[str]:
    yield "hello world"
    yield "my name is john smith"
    yield "i am a software engineer"


gen_obj: Generator[str] = my_gen()

for one_string in gen_obj:
    print(one_string)
````

* The **return** statement is used to **immediately exit** a function and **return a value** to the caller.
* The **yield** statement is used in a **generator function** to **produce a value** to the caller,
but it **doesn't immediately terminate** the function's execution.
* Instead, it **temporarily suspends** the function and saves its state.
* A **function with yield** is a **generator** function, while a **function** with **return** is a **regular** function.
* When a **generator** function is **called**, it **doesn't execute** immediately.
* Instead, it **returns a generator object** which can be **iterated** over.
* Generators are used to **create iterators**, which allow you to **iterate over values** **one by one** without loading the **entire sequence** into memory.
* Generator functions maintain their state between calls, allowing them to **resume execution** and **produce more values** over multiple calls.

````python
# short hand way to create generators through comprehension
import sys
from typing import TYPE_CHECKING

if TYPE_CHECKING:
    from collections.abc import Generator

n: int = 100_000

my_list: list[int] = [i*i for i in range(1, n)]
print(type(my_list))
print(f"size of my_list: {sys.getsizeof(my_list)} bytes.")

# its not tuple comprehension.() doesn't make a tuple.
my_gen: Generator[int] = (i*i for i in range(1, n))
print(type(my_gen))
print(f"size of my_gen: {sys.getsizeof(my_gen)} bytes.")
````
