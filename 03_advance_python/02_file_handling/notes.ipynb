### Course Notes:-

- File handling is a crucial aspect of programming that allows us to read from and write to files on our computer.
- In Python, we can use the built-in `open()` function to work with files. The `open()` function takes two arguments: the file name and the mode in which we want to open the file (e.g., 'r' for reading, 'w' for writing, 'a' for appending).
- When we open a file, we get a file object that we can use to perform various operations on the file, such as reading its contents, writing to it, or closing it when we're done.
- It's important to close the file after we're done with it to free up system resources. We can use the `close()` method to close the file, or we can use a `with` statement to automatically close the file when we're done.
- To read the contents of a file, we can use the `read()` method, which reads the entire contents of the file as a string.
- We can also use the `readline()` method to read one line at a time, or the `readlines()` method to read all lines into a list.
- When writing to a file, we can use the `write()` method to write a string to the file, or the `writelines()` method to write a list of strings to the file.
- When working with files, it's important to handle exceptions that may occur, such as `FileNotFoundError` when trying to open a file that doesn't exist, or `OSError` when there are issues with reading or writing to the file.
- So, We should use try-except blocks to catch these exceptions and handle them gracefully in our code.
- For smaller applications use **file_handling**, but for larger applications, it's better to use a **database** to store and manage data efficiently.

### File-Handling :-

- **What is file handling?**
  - open a file - performing some operations on it
  - close the file
- **What is a file?**
  - Files are named locations on disk to store information.
  - They are used to store data permanently.
  - Data stored in files are non-volatile memory
  - we can retrieve data whenever required
- **Types of files**
  - Text Files
    - store data in the form of characters. it is used to store data and strings.
  - Binary Files
    - store data in the form of bytes(group of 8 bits). it is used to store images, videos, pdf, audio files, etc.

**Syntax:-**

```python
f = open(filename, mode='r', buffering, encoding=None,
errors=None, newline=None, closefd=True)
```

#### file object parameters:-

1. it provides a file handler(f), which is used to perform operations, by using file object functions.
2. **f(obj)** - refers to the beginning of the file.
3. **file name** and **file mode** should be **string**
4. **buffering**:- Positive integer value, use to set buffer size of the file. 
    1. In Text mode, buffer size should be 1 or more then 1
    2. In Binary mode, buffer size can be Zero(0)
    3. Default Size is :- 4096-8198 bytes, it depends on the operating system.(if we didn't mention buffering size, then it will vary between these values.)
   
   For example, we have a file with 1GB of memory size, if we need to fetch data from the file, it is not good to call all file at onces, or not its efficient. so we provide a buffer values, that divide the file data into that provided size into the system buffer memory. if we give buffer size(n) of 10, then 1GB file size will divided into that chunk size. so at one time, only that 1 number of chuck will be accessed and processed.

5. **Encoding:-** it is used to decode and encode files.
    1. should be used in text mode, it is not used in binary mode.
    2. default value depends on OS.
    3. for windows, default encoding is 'cp1252'. we can specify encoding as 'utf-8' to avoid encoding issues.
    4. for linux, default encoding is 'utf-8'
6. **Errors:-** it is used to handle encoding and decoding errors. it is used in text mode, it is not used in binary mode.
    1. default value is 'strict', it raises a UnicodeDecodeError when an error occurs.
    2. 'ignore' - ignores the error and continues with the next character.
    3. 'replace' - replaces the error with a replacement character (e.g., '?').
    4. 'backslashreplace' - replaces the error with a backslash escape sequence.
    5. 'xmlcharrefreplace' - replaces the error with an XML character reference.
7. **Newline:-** it is used to handle newlines in text mode. it is not used in binary mode.
    1.  default value is None, it means that newlines are handled in a platform-specific way.
    2.  we will set it to the default value, because its different for every file type. Such as for csv files, it is '\r\n', for txt files, it is '\n', for mac files, it is '\r'.
    3.  if we set newline to '', it means that newlines are not translated, and they are returned as they are in the file.
    4.  if we set newline to '\n', it means that all newlines are translated to '\n'.
    5.  if we set newline to '\r', it means that all newlines are translated to '\r'.
    6.  if we set newline to '\r\n', it means that all newlines are translated to '\r\n'.

#### Closing a file:-

After performing operations, we have to close a file.

`close()` :- function used to close a file.

**Syntax :-**

```python
file_handler.close()
```

What happens when we close a file?

File object is deleted from memory and file is no more accessible unless we open it again.

What happens when we do not close a file ?:-

After program execution python garbage collector will destroy file object and closes file automatically.
Note :- Don't rely on garbage collector.

#### reading file in read-only mode - old basic way

````python
from typing import TYPE_CHECKING

if TYPE_CHECKING:
    from io import TextIOWrapper

my_file: TextIOWrapper = open(
    "data/data.txt", mode="r", buffering=10, encoding="utf-8")
print(my_file.read())  # read file content
print("===" * 25)
print(my_file)  # prints file object information

# close the open file after operations:
my_file.close()
````

#### File object methods:-

**readable() :-**

This method is used to check whether file is readable or not.
- True :- if file is readable.
- False :- if file is not readable.

**writable() :-**

This method is used to check whether file is writable or not.
- True :- if file is writable.
- False :- if file is not writable.

````python
from typing import TYPE_CHECKING

if TYPE_CHECKING:
    from io import TextIOWrapper

my_file: TextIOWrapper = open(
    # we removed mode='r' because it's default value is set to 'r', so need to mention it.
    "data/data.txt", mode="r+", buffering=10, encoding="utf-8")

print(my_file.readable())
print(my_file.writable())
my_file.close()
````

#### Mode of opening file:-

when you open a file for operations, you have to specify mode. **mode specifies the purpose of opening file**. There are two types of modes :-
1. Text modes
2. Binary modes.

**Text mode :-**

- when you open a file in text modes, python treats it's content as "str" type.
- When you get data from a text mode file, python first decodes the raw bytes using either a platform dependent encoding or specified encoding.

1. **read only mode** - 'r' - it is **default value** of mode. Open file for reading only. If the file doesn't exist, it will show `FileNotFoundError`.
2. **read and write mode** - 'r+' - it is used to open for reading and then writing(append) data to a file.
3. **write-only mode** - 'w' - it is used to write data to a file. Open file for writing only. If the file doesn't exist, it will create
a file.
4. **write and read mode** - 'w+' - it is used to open for writing and then reading data from the file.
5. **append mode** - 'a' - it is used to write data to a file. Open file for appending only. If the file doesn't exist, it will create a file.
6. **append and read mode** - 'a+' - it is used to open for appending and then writing data to the file.
7. **exclusive creation mode** - 'x' - Open for exclusive creation for writing.The specified file must not be available. It creates a new file and then we write data into it. If file exists , it will give an error.

**Binary mode :-**

- When you open a file for binary mode, python uses the data **without any encoding**.
- Binary mode file reflects the raw data directly in the file.
- Python treats file contents as "bytes" type. These modes are used while dealing with non-text files like images, audios, videos etc.

- rb - Open for reading in binary mode.(same as text 'b')
- wb - Open for writing in binary mode.(same as text 'w')
- ab - Open for appending.(same as text 'a').
- xb - Open for exclusive creation and writing.(same as 'x')
- rb+ - Open for read and then write in binary.
- wb+ - Open for writing and then reading in binary.
- ab+ - Open for append and then read in binary

### Reading data from file :-

#### better way to read all file - using with, try-except block

````python
file_path: str = "data/data.txt"

# Reading - First time
try:
    print("📖 Reading file...")

    with open(file_path, encoding="utf-8") as file:
        print(file.read())
    # with helps to close file automatically
except FileNotFoundError:
    print("The file 'employees.csv' was not found.")
except OSError as e:
    print(f"An error occurred while reading the file: {e}")
````

#### better way to read single line from file - using with, try-except block

````python
file_path: str = "data/data.txt"

# Reading - First time
try:
    print("📖 Reading file...")

    with open(file_path, encoding="utf-8") as file:
        print(file.readline())
    # with helps to close file automatically
except FileNotFoundError:
    print("The file 'employees.csv' was not found.")
except OSError as e:
    print(f"An error occurred while reading the file: {e}")
````

#### better way to read N(chars) from file - using with, try-except block

````python
file_path: str = "data/data.txt"

# Reading - First time
try:
    print("📖 Reading file...")

    with open(file_path, encoding="utf-8") as file:
        # size argument is used to read specific number of characters from the line. If size is not provided, it reads the whole line.
        print(file.readline(7))
    # with helps to close file automatically
except FileNotFoundError:
    print("The file 'employees.csv' was not found.")
except OSError as e:
    print(f"An error occurred while reading the file: {e}")
````

#### better way to read file as a List - using with, try-except block

````python
file_path: str = "data/data.txt"

# Reading - First time
try:
    print("📖 Reading file...")

    with open(file_path, encoding="utf-8") as file:
        # return list of all lines in the file. Each line is a string element in the list.
        print(file.readlines())
    # with helps to close file automatically
except FileNotFoundError:
    print("The file 'employees.csv' was not found.")
except OSError as e:
    print(f"An error occurred while reading the file: {e}")
````

#### `tell()`

- This method is used to find the **current position** of a **file pointer** from the **beginning of the file**.
- Position starts from 0. It is just like indexing in string.

````python
file_path: str = "data/data.txt"

# Reading - First time
try:
    print("📖 Reading file...")

    with open(file_path, encoding="utf-8") as file:
        print(file.tell())
        print(file.read(7))
        print(file.tell())
    # with helps to close file automatically
except FileNotFoundError:
    print("The file 'employees.csv' was not found.")
except OSError as e:
    print(f"An error occurred while reading the file: {e}")
````

#### `seek()`

- This method is used to change the position of file pointer.
- Remember, file pointer position is always counted from the beginning.

#### Reading `csv` file :-

What is dialect in csv file ?:-
- Dialect includes parameters which specifies conventions or information to parse the csv file.
- It contains delimiter information, header information, quoting character etc
- You can set dialect using `register_dialect()` function.

````python
import csv

file_path: str = "data/employees.csv"

# Reading - First time
try:
    print("📖 Reading file...")

    with open(file_path, encoding="utf-8") as file:
        employee_data = csv.reader(file)
        for employee in employee_data:
            print(employee)
except FileNotFoundError:
    print("The file was not found.")
except OSError as e:
    print(f"An error occurred while reading the file: {e}")
````

### Writing to File

#### writing to file - data overwrite - using with, try-except block

````python
file_path: str = "data/data.txt"

try:
    # Writing - data overwrite the file content
    print("Writing to file...")
    user_input: str = input("Please enter your name: ")

    with open(file_path, "w", encoding="utf-8") as file:
        file.write(user_input)
    # file automatically closes

    # Reading - Second Time
    print("\n📖 Reading file (updated)...")
    with open(file_path, encoding="utf-8") as file:
        print(file.read())
    # file automatically closes

except FileNotFoundError:
    print("The file 'employees.csv' was not found.")
except OSError as e:
    print(f"An error occurred while reading the file: {e}")
````

#### writing to file - data on new line - using with, try-except block

````python
file_path: str = "data/data.txt"

try:
    print("Writing to file...")
    user_input: str = input("Please enter your name: ")

    with open(file_path, "a", encoding="utf-8") as file:
        file.write(user_input + "\n")

    print("\n📖 Reading file (updated)...")
    with open(file_path, encoding="utf-8") as file:
        print(file.read())

except FileNotFoundError:
    print("The file 'employees.csv' was not found.")
except OSError as e:
    print(f"An error occurred while reading the file: {e}")
````

#### writing to file - insert iterable - using with, try-except block

````python
file_path: str = "data/data.txt"

try:
    print("Writing to file...")
    user_names: list[str] = ["salman", "ahmed", "ali", "ramzan"]

    with open(file_path, "a", encoding="utf-8") as file:
        file.writelines([name + "\n" for name in user_names])

    print("\n📖 Reading file (updated)...")
    with open(file_path, encoding="utf-8") as file:
        print(file.read())

except FileNotFoundError:
    print("The file 'employees.csv' was not found.")
except OSError as e:
    print(f"An error occurred while reading the file: {e}")
````

````python
file_path: str = "data/data.txt"

try:
    print("Writing to file...")
    user_names: tuple[str, ...] = ("jawad", "bala", "usman", "cherkutie")

    with open(file_path, "a", encoding="utf-8") as file:
        file.writelines([name + "\n" for name in user_names])

    print("\n📖 Reading file (updated)...")
    with open(file_path, encoding="utf-8") as file:
        print(file.read())

except FileNotFoundError:
    print("The file 'employees.csv' was not found.")
except OSError as e:
    print(f"An error occurred while reading the file: {e}")
````

#### a+ mode - using with, try-except block - Write data to a file, then setting cursor to the beginning of file and then read the file content.

````python
file_path: str = "data/data.txt"

try:
    print("Writing to file...")
    user_names: tuple[str, ...] = ("Usman", "bala", "Orhan", "Cherkutie")

    # this will append data at the end of file
    with open(file_path, "a+", encoding="utf-8") as file:
        file.writelines([name + "\n" for name in user_names])
        file.seek(0)  # Move the file pointer to the beginning
        print(file.read())

except FileNotFoundError:
    print("The file 'employees.csv' was not found.")
except OSError as e:
    print(f"An error occurred while reading the file: {e}")
````
