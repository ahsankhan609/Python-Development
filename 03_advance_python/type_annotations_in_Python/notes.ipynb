## why type annotation?

we used them just to make sure, that we use them in the right context in which we are passing it.

for example if we are creating a variable that is of **type string**, then we should use `name:str='hello world'`. it will prevent use to passing any other type, such as **integer**.

it doesn't mean that our code should not run, python has no issue with type annotations, it will, but the editor will complaint about the **type** that we pass in. It will create problem later.

So type annotations will help us in the debugging later.

### Union Types:

so how we annotate if our `var` should be of type `str` and `int` at a same time or have the possibility of having both.

```python
var1: str | int = "python"
var2: str | int = 10

print(var1, var2)  # Output: python 10
```

```python
# Python's grammar doesn't support applying a single type annotation to multiple targets in tuple unpacking.
# The type checker would have no way to know which variable gets which type hint if you tried something like:

var1, var2: str | int = "python" , 10
# var2 : str | int = "python"

print(var1, var2)
```

#### Using a Type Alias (If You Want Reusability) - before Python 3.10

```python
from typing import Union

StringOrInt = str | int  # Python 3.10+

var1: StringOrInt = "python"
var2: StringOrInt = 10

print(var1, var2)  # Output: python 10
```

```python
from typing import Union

var3: Union[str, int] = 'python'
print(var3, type(var3))
```

### Optionals

```python
# here we can understand it with an example. for example, we have a program that greet user, when he/she provide his/her name.
# but he/shee didn't provide their names, then we should welcome the guest user.
# so the guest user is of type Optionals


def greet_user(username:str | None = None) -> None:
    """
    it welcomes the user if name is provided, else it welcomes as guest user.

    Raise:
        - Return AttributeError if integer value given

    Args:
        username (str, optional): accept name as string. Defaults to None.

    Examples:
        greet_user('john')
    """
    person:str | None = 'Guest user'

    if username is None or username == '':
        print(f"Welcome! {person}")
    else:
        print(f"Hello!", username.title())

try:
    greet_user()
except AttributeError as ae:
    print(f"Please provide your name in string. {ae}")
except Exception as e:
    print("Please provide your correct name. {e}")
```

### type annotation in Classes

```python
class Fruits:
    """
    A class to represent a fruit with its name and weight.

    Attributes:
        fruit_name (str): The name of the fruit.
        fruit_weight_in_grams (float): The weight of the fruit in grams.
    """
    def __init__(self, fruit_name:str, fruit_weight_in_grams:float):
        """
        Initialize a Fruits object with a name and weight.

        Args:
            fruit_name (str): The name of the fruit.
            fruit_weight_in_grams (float): The weight of the fruit in grams.
        """
        self.fruit_name = fruit_name
        self.fruit_weight_in_grams = fruit_weight_in_grams

    def __str__(self):
        """
        Return a string representation of the Fruits object.

        Returns:
            str: A formatted string showing the fruit name and weight in grams.
        """
        return f"({self.fruit_name}) weight ({self.fruit_weight_in_grams}) grams."
```

```python
# Create an instance of Fruits class with an orange
orange:Fruits = Fruits('Orange', 253.63)
print(orange)
```

```python
class Apple(Fruits):
     """
    A child class to represent an apple, inheriting from the Fruits class.

    This class inherits all attributes and methods from the Fruits class
    and can be extended with apple-specific properties and behaviors.
    """
     pass

# Create an instance of Apple class (child class of Fruits)
apple:Fruits = Apple('fuji red apples', 253.63)
print(apple)
```

### Return Types

it is always a good practice to provide a return type when we construct any function. it will help other developers to understand what type of data our function should return.

but if our function doesn't return anything or jsut do the process, then we should return `None` to explicitly tell user.

```python
def add_number(a:int, b:int) -> int:
    return a + b

print(5+6)
```

```python
# let's say our function return nothing then we do like this:

def running() -> None:
    print("running in the morning...")

running()
```

```python
# another example

def fetch_user_name_by_id(user_id:int) -> str | None:
    users = {
        1 : 'mario',
        2 : 'fuji',
        3 : None,
    }
    return users.get(user_id)

print(fetch_user_name_by_id(1))
print(fetch_user_name_by_id(3))
```

- so, we can import it from the library explicitly
