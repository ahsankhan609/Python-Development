## why type annotation?

we used them just to make sure, that we use them in the right context in which we are passing it.

for example if we are creating a variable that is of **type string**, then we should use `name:str='hello world'`. it will prevent use to passing any other type, such as **integer**.

it doesn't mean that our code should not run, python has no issue with type annotations, it will, but the editor will complaint about the **type** that we pass in. It will create problem later.

So type annotations will help us in the debugging later.

### Optionals

```python
# here we can understand it with an example. for example, we have a program that greet user, when he/she provide his/her name.
# but he/shee didn't provide their names, then we should welcome the guest user.
# so the guest user is of type Optionals


def greet_user(username:str | None = None) -> None:
    """
    it welcomes the user if name is provided, else it welcomes as guest user.

    Raise:
        - Return AttributeError if integer value given

    Args:
        username (str, optional): accept name as string. Defaults to None.

    Examples:
        greet_user('john')
    """
    person:str | None = 'Guest user'

    if username is None or username == '':
        print(f"Welcome! {person}")
    else:
        print(f"Hello!", username.title())

try:
    greet_user()
except AttributeError as ae:
    print(f"Please provide your name in string. {ae}")
except Exception as e:
    print("Please provide your correct name. {e}")
```

### type annotation in Classes

```python
class Fruits:
    """
    A class to represent a fruit with its name and weight.

    Attributes:
        fruit_name (str): The name of the fruit.
        fruit_weight_in_grams (float): The weight of the fruit in grams.
    """
    def __init__(self, fruit_name:str, fruit_weight_in_grams:float):
        """
        Initialize a Fruits object with a name and weight.

        Args:
            fruit_name (str): The name of the fruit.
            fruit_weight_in_grams (float): The weight of the fruit in grams.
        """
        self.fruit_name = fruit_name
        self.fruit_weight_in_grams = fruit_weight_in_grams

    def __str__(self):
        """
        Return a string representation of the Fruits object.

        Returns:
            str: A formatted string showing the fruit name and weight in grams.
        """
        return f"({self.fruit_name}) weight ({self.fruit_weight_in_grams}) grams."
```

```python
# Create an instance of Fruits class with an orange
orange:Fruits = Fruits('Orange', 253.63)
print(orange)
```

```python
class Apple(Fruits):
     """
    A child class to represent an apple, inheriting from the Fruits class.

    This class inherits all attributes and methods from the Fruits class
    and can be extended with apple-specific properties and behaviors.
    """
     pass

# Create an instance of Apple class (child class of Fruits)
apple:Fruits = Apple('fuji red apples', 253.63)
print(apple)
```

### Sequences

it is somehow similiar to the Iterables.

```python
from typing import Sequence, Iterable

# Set's are an un-ordered collection of elements. so we can't access element with index
sample_set: set[str] = {'first_set', 'second_set', 'third_set'}

# List are an ordered collection of elements. so we can access element with index
sample_list: list[str] = ['first_list', 'second_list', 'third_list']

def get_first_element(sequence: Sequence[str]) -> str | int:
    """
    - this function takes a sequence of elements and returns the first element in the sequence.
    - it works with ordered type of elements such as Lists.
    - questions comes in our mind that we can mention Iterable as Type, and it will works fine. But actually it works with Loops only.
    - So we only use Sequence type for the elements

    Args:
        sequence (Sequence[str]): The sequence of elements.

    Raises:
        It will raise `TypeError` if we try to pass un-ordered elements. Such as Set of elements
        >>> TypeError: 'set' object is not subscriptable

    Returns:
        str: The first element in the sequence.

    Example:
          >>> print(get_first_element(['first_list_element', 'second_list_element', 'third_list_element']))
          >>> 'first_list_element'
    """
    return sequence[0] if sequence else -1

try:
    print(get_first_element(sample_list)) # because Sets are un-ordered, that's why it creates warning
    print(get_first_element(sample_set)) # Throw TypeError : because Sets are un-ordered
except Exception as e:
    print(e)
```

```python
"""
Module: sequence_demo.py

Sequence vs Iterable - Type Annotations in Python
--------------------------------------------------
This module demonstrates the difference between:
    - Sequence (ordered, indexable: list, tuple, str)
    - Iterable (loop-only: set, generator, etc.)
"""

from typing import Sequence, Iterable


# ──────────────────────────────────────────────
# Sample Collections
# ──────────────────────────────────────────────

# Sets are UN-ORDERED — no index access possible
sample_set: set[str] = {"first_set", "second_set", "third_set"}

# Lists are ORDERED — index access is possible
sample_list: list[str] = ["first_list", "second_list", "third_list"]

# Tuples are ORDERED — index access is possible (Sequence!)
sample_tuple: tuple[str, ...] = ("first_tuple", "second_tuple", "third_tuple")

# Strings are also a Sequence — every character is indexed
sample_string: str = "Python"


# ──────────────────────────────────────────────
# Function 1: Using Sequence (Correct Approach)
# ──────────────────────────────────────────────

def get_first_element(sequence: Sequence[str]) -> str:
    """
    Return the first element from any ordered, indexable collection.

    This function accepts any type that implements the Sequence protocol,
    meaning it must support:
        - len()        → to check if it's empty
        - indexing[]   → to access elements by position

    Valid types  : list, tuple, str, range, bytes
    Invalid types: set, dict (keys), generator → no index support

    Args:
        sequence (Sequence[str]): An ordered, indexable collection of strings.

    Returns:
        str: The first element of the sequence.

    Raises:
        ValueError: If the sequence is empty.
        TypeError : If a non-subscriptable type (e.g. set) is passed at runtime.

    Examples:
        >>> get_first_element(["alpha", "beta", "gamma"])
        'alpha'
        >>> get_first_element(("x", "y", "z"))
        'x'
        >>> get_first_element("Python")
        'P'
    """
    if not sequence:
        # Returning -1 for empty is bad practice — raise a clear exception
        raise ValueError("Cannot get first element: sequence is empty.")

    return sequence[0]  # index[0] works because Sequence guarantees ordering


# ──────────────────────────────────────────────
# Function 2: Using Iterable (WHY it's not enough)
# ──────────────────────────────────────────────

def print_all_elements(iterable: Iterable[str]) -> None:
    """
    Iterate over any collection and print each element.

    Iterable only guarantees that we can loop over it (via __iter__).
    It does NOT guarantee:
        - ordering
        - index access (no sequence[0])
        - len() support

    This is why Iterable alone is insufficient when you need index access.

    Args:
        iterable (Iterable[str]): Any iterable collection of strings.

    Returns:
        None
    """
    for element in iterable:  # works with list, set, tuple, generator — anything iterable
        print(f"  → {element}")


# ──────────────────────────────────────────────
# Demonstration
# ──────────────────────────────────────────────

def main() -> None:
    """Run all demonstrations to explain Sequence vs Iterable."""

    print("=" * 55)
    print(" SEQUENCE — Ordered & Indexable")
    print("=" * 55)

    # list → Sequence ✅
    print("\n[list] get_first_element:")
    print(f"  Result: {get_first_element(sample_list)}")

    # tuple → Sequence ✅
    print("\n[tuple] get_first_element:")
    print(f"  Result: {get_first_element(sample_tuple)}")

    # str → also a Sequence! each char is indexed ✅
    print("\n[str] get_first_element (string is a Sequence of chars):")
    print(f"  Result: {get_first_element(sample_string)}")

    print("\n" + "=" * 55)
    print(" ITERABLE — Loop-only (no index guarantee)")
    print("=" * 55)

    # set → Iterable ✅ (can loop), but NOT a Sequence ❌ (no index)
    print("\n[set] print_all_elements (loop works fine on set):")
    print_all_elements(sample_set)

    print("\n" + "=" * 55)
    print(" RUNTIME ERROR — Passing set to Sequence function")
    print("=" * 55)

    # Passing set to a Sequence function → TypeError at runtime
    # Type checkers (mypy / pylance) also warn at STATIC analysis time
    print("\n[set] get_first_element (will raise TypeError):")
    try:
        result = get_first_element(sample_set)  # type: ignore[arg-type]
        print(result)
    except TypeError as te:
        # set does not support index access → 'set' object is not subscriptable
        print(f"  ❌ TypeError caught: {te}")

    print("\n" + "=" * 55)
    print(" EMPTY SEQUENCE — Proper error handling")
    print("=" * 55)

    print("\n[] get_first_element on empty list (will raise ValueError):")
    try:
        result = get_first_element([])
        print(result)
    except ValueError as ve:
        # We raise ValueError explicitly for empty sequences
        print(f"  ❌ ValueError caught: {ve}")


if __name__ == "__main__":
    main()
```

### Callables

```python
from typing import Callable
"""
When we need to call A function in our main function to do something is called Callable. So we explicitly mention it will the `Callable` type of annotation.
We mention other function `input types` and what it returns, in our main function as mentioned in below exmaple.
"""
def print_it(user_text: str, print_func: Callable[[str], None]) -> None:
    """
    - it takes a print_func(other function) as argument.
    - print_func(other function) is called Callable, takes str and return None.
    - the print_it(main function) returns None.
    - print_it(main function) just prints the output of input print_func(other function).
    """
    print_func(user_text)


def loud_print(user_text: str) -> None:
    """
    this is an example func that we will use inside the print_it (main function). we mentioned its characteristics in print_it function. such as it take string as input and return None.
    """
    print(f'{user_text. upper ()}!')


def silent_print(user_text: str) -> None:
    """
    this is an example func that we will use inside the print_it function. we mentioned its characteristics in print_it function. such as it take string as input and return None.
    """
    print(f'({user_text.lower()})')
```

### Protocols

it defines the structure of what a Class should look like. For example:- if we have a Printer class, we are going to want it to have a certain functionality, and every class that conforms to the printer protocol must have this functionality to be considered as a Printer.

```python
from typing import Protocol

class Printer(Protocol):
    """
    Defines our Protocol Class
    """
    def print_me(self, magazine_name:str) -> None:
        pass # we can use ... as placeholder as well

    def copies(self, magazine_name: str, copies: int) -> None:
        pass
```

```python
class LazerPrinter:
    """
    Defines our Printer class that inherit protocols of Parent-> Printer Class
    """
    def __init__(self, printer_name: str, printer_version: int) -> None:
        self.printer_name = printer_name
        self.printer_version = printer_version

    # Protocol Function
    def print_me(self, magazine_name: str) -> None:
        print(f'{self.printer_name} ({self.printer_version}) is printing: "{magazine_name}".')
    # Protocol Function
    def copies(self, magazine_name: str, copies: int) -> None:
        print(f'{self.printer_name} ({self.printer_version}) is making {copies} copies of: "{magazine_name}".')

    # Own Function
    def faxing(self, magazine_name: str, city_name:str) -> None:
        print(f"{self.printer_name} is faxing {magazine_name} to {city_name}.")
```

```python
class InkJetPrinter:
    """
    Defines our Printer class that inherit protocols of Parent-> Printer Class
    """
    def __init__(self, printer_name: str, printer_version: int) -> None:
        self.printer_name = printer_name
        self.printer_version = printer_version

    # Protocol Function
    def print_me(self, magazine_name: str) -> None:
        print(f'{self.printer_name} ({self.printer_version}) is printing: "{magazine_name}".')
    # Protocol Function
    def copies(self, magazine_name: str, copies: int) -> None:
        print(f'{self.printer_name} ({self.printer_version}) is making {copies} copies of: "{magazine_name}".')

    # Own Function
    def faxing(self, magazine_name: str, city_name:str) -> None:
        print(f"{self.printer_name} is faxing {magazine_name} to {city_name}.")
```

```python
# construct instance of the Printer class
lp:Printer = LazerPrinter('HP LazerPrinter', 5400)
inkj:Printer = InkJetPrinter('HP InkJetPrinter', 5300)
```

```python
def use_printer(printer_name:Printer, magazine_name:str) -> None:
    """
    define a function to use LazerPrinter class Instance.
    """
    printer_name.print_me(magazine_name)
    printer_name.copies(magazine_name, 500)
    printer_name.faxing(magazine_name, 'UK')

use_printer(lp, 'Python Geek')
print("==" * 50)
use_printer(inkj, 'Python Mojo')
```
