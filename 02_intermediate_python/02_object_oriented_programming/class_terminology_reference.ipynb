**Class Terminology Reference**

| Term               | Also Known As                    | Definition                                                                                                  | Notation                                  | Example                                    |
| ------------------ | -------------------------------- | ----------------------------------------------------------------------------------------------------------- | ----------------------------------------- | ------------------------------------------ |
| Class              | Blueprint, Template              | The structure/definition for creating objects                                                               | class ClassName:                          | class Car:                                 |
| Instance           | Object                           | A concrete realization of a class                                                                           | obj = ClassName()                         | toyota = Car(...)                          |
| Instance Attribute | Instance Variable, Property      | Variables that belong to a specific object instance; each instance has its own copy                         | self._attribute or self.attribute         | self._name = "Camry"                       |
| Class Attribute    | Class Variable, Static Variable  | Variables that belong to the class itself; shared by all instances                                          | Declared before __init__, outside methods | COMPANY_NAME = "Toyota"                    |
| Instance Method    | Method, Function                 | Function that operates on instance data; has access to self                                                 | def method(self):                         | def get_name(self):                        |
| Class Method       | Classmethod                      | Function that operates on class data; has access to cls instead of self                                     | @classmethod def method(cls):             | @classmethod def get_company_details(cls): |
| Static Method      | Staticmethod, Utility Function   | Helper function logically related to class but independent of instance/class data; no access to self or cls | @staticmethod def method():               | @staticmethod def calculate_car_age(year): |
| Constructor        | Initializer                      | Special method that runs when an object is created; initializes instance attributes                         | def __init__(self, ...):                  | def __init__(self, name, year):            |
| Property (Getter)  | Getter                           | Method that retrieves/returns an attribute value using @property decorator                                  | @property def attr(self):                 | @property def name(self):                  |
| Setter             | Property Setter                  | Method that sets/modifies an attribute value with validation using @property.setter                         | @attr.setter def attr(self, value):       | @name.setter def name(self, value):        |
| Magic Method       | Dunder Method, Special Method    | Built-in methods with double underscores that define special behavior                                       | def __method__(self):                     | def __str__(self):, def __repr__(self):    |
| Encapsulation      | Data Hiding, Protected Attribute | Practice of using underscore prefix to indicate attributes should not be accessed directly from outside     | self._attribute                           | self._name (protected)                     |
| Inheritance        | Subclassing, Extending           | When a class extends another class to reuse/modify its behavior                                             | class Child(Parent):                      | class ElectricCar(Car):                    |
| Override           | Method Overriding                | Redefining a parent class method in a child class                                                           | In child class method definition          | Child's __init__ overrides parent's        |

**Quick Lookup by Use Case**

| I Want to...                           | Use This                           |
| -------------------------------------- | ---------------------------------- |
| Store data unique to each object       | Instance Attribute (self._attr)    |
| Store data shared by all instances     | Class Attribute (ATTR = value)     |
| Access/retrieve object data safely     | Property with @property (getter)   |
| Modify object data with validation     | Setter with @property.setter       |
| Perform actions using object data      | Instance Method (def method(self)) |
| Perform actions using class data       | Class Method (@classmethod)        |
| Create a helper function for the class | Static Method (@staticmethod)      |
| Initialize an object when created      | Constructor (def __init__)         |
| Define custom print behavior           | Magic Method (def __str__)         |
| Hide attributes from direct access     | Encapsulation (use _ prefix)       |

**Visual Structure Example**

```python
from typing import Literal


class Car:                                          # CLASS
    COMPANY_NAME = "Toyota"                        # CLASS ATTRIBUTE

    # CONSTRUCTOR (special instance method)
    def __init__(self, name: str, year: int) -> None:
        self._name: str = name                          # INSTANCE ATTRIBUTE
        self._year: int = year                          # INSTANCE ATTRIBUTE

    @property
    def name(self) -> str:                                # PROPERTY/GETTER
        return self._name

    @name.setter
    def name(self, value: str) -> None:                         # SETTER
        self._name = value

    def get_details(self) -> str:                         # INSTANCE METHOD
        return f"{self.name} ({self._year})"

    @classmethod
    # CLASS METHOD
    def get_company(cls) -> Literal['Toyota']:
        return cls.COMPANY_NAME

    @staticmethod
    def is_vintage(year):                          # STATIC METHOD
        return 2024 - year >= 25

    def __str__(self):                             # MAGIC METHOD/DUNDER
        return self.get_details()
```

```python
# INSTANCE/OBJECT creation
toyota = Car("Camry", 2020)                        # INSTANCE


# Accessing instance attribute via property
print(toyota.name)                                 # getter


# Modifying instance attribute via property
toyota.name = "Corolla"                            # setter


# Calling instance method
print(toyota.get_details())


# Calling class method
print(Car.get_company())


# Calling static method
print(Car.is_vintage(1995))
```
