### Classes & Objects

```python
class Lamp:
    # class variables - defined outside the methods
    COMPANY_COUNTRY:str = 'USA'

    def __init__(self, model_name:str, color:str) -> None:
        self.model_name:str = model_name # self.model_name is called instance variables, because they are referring to the self
        self.color:str = color

    def turning_on(self) -> None:
        print(f"{self.model_name} is turned on.")

    def turning_off(self) -> None:
        print(f"{self.model_name} is turned off.")

    def describe_me(self) -> None:
        print(f"{self.model_name} is of {self.color}. This {self.model_name}.")
```

```python
samsung:Lamp = Lamp('Samsung', 'Red')
samsung.turning_on()
samsung.describe_me()
samsung.turning_off()

hitachi:Lamp = Lamp('Hitachi', 'Blue')
hitachi.turning_on()
hitachi.describe_me()
hitachi.turning_off()
```

### geters and setters

```python
class Fruit:
    def __init__(self, name:str) -> None:
        self._name: str = name # _name is a private variable, it just naming convention

    # define getter method for our class
    @property
    def get_fruit_name(self):
        print(f"Fruit name is: {self._name}.")
        return self._name

    # define setter method for our class
    @get_fruit_name.setter
    def set_fruit_name(self, updated_name:str):
        print(f"Fruit name was: {self._name}, now changed to {updated_name}.")
        self._name = updated_name
```

```python
apple:Fruit = Fruit('Apple')
apple.get_fruit_name # if we use @property decorator, we will access it without (), because @property handles rest

# set new name
apple.set_fruit_name = 'Banana'
print(apple.get_fruit_name)
```
