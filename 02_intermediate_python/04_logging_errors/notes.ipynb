### What is logging?

1. logging is used to track events that occur in software when it runs.

2. It is used in development, debugging, testing etc

3. Logging is useful to track the error or exception or information.

4. logging module can be used with frameworks like django,flask etc

### What is if there is no logging

1. RCA :- Root Cause Analysis, will be difficult to find root cause.
2. it may take long time to find root cause.
3. Difficulty in debugging and testing

### What problems software can face?

1. Exceptions present in that programming language. Such as ZeroDivisionError, NameError, etc.
2. Network issues and connection issues. Such as ConnectionError, TimeoutError, etc.
3. Access denied issues like permission issues. Such as PermissionError, etc.
4. Server down because of maintenance activity. 
5. File not found

#### Logging levels

1. **DEBUG** :- Release information useful for debugging purpose.
2. **INFO** :- Provides the information regarding that things are working as we want.
3. **WARNING** :- used to warn that something happened unexpectedly.
4. **ERROR** :- used to inform when we are in some serious trouble.
5. **CRITICAL** :- used to inform when we are in some serious trouble.
