**Advantages of recursion**
- Clean code
- Complex problems can be solved

**Disadvantages of recursion**

- Hard to debug
- Not memory efficient

When a function calls itself,inside same function, then that function is called as `recursive function` and `process` is called as `recursion`.

```python
# https://youtu.be/ivl5-snqul8?si=NdiMp20QHf3v62X-

# recursion = a function that calls itself from within
# helps to visualize a complex problem into basic steps,
# problems can be solved more easily iteratively or recursively
# iterative = faster, complex
# recursive = slower, simpler

# ITERATIVE
def itr_walk(steps: int) -> None:
    for step in range(1, steps + 1):
        print(f"You take step #{step}")

# RECURSIVE


def rec_walk(steps: int) -> None:
    if steps == 0:
        return
    rec_walk(steps - 1)
    print(f"You take step #{steps}")


itr_walk(10)
print("=============================Now walk Recursively==========================================")
rec_walk(10)
```

```python
import sys
print(sys.getrecursionlimit())

i = 0


def demo_func():
    global i
    i += 1
    print(i, "- Demo Function Recursion runs 1000 times. Limit set by python.")
    demo_func()


# demo_func()  # RecursionError: maximum recursion depth exceeded
```

**Basic types of recursion :-**
- Direct Recursion - when a funciton calles itself
- Indirect Recursion - a function calls another function which then calls the first function again.

```python
# direct recursion

n: int = int(input("Enter the value of n: "))


def natural(n: int) -> None | int:  # -> None | Any:
    if n == 0:
        return
    print(n)
    return natural(n-1)


natural(n)
```
