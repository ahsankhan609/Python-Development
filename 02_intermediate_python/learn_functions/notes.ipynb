#### Scope in programming

In programming, the term "scope" refers to the region of the code where an identifier can be accessed.

#### What is an identifier ?

an identifier is a name using which we can identify a particular object. Objects can be variable, function name, class name etc.

In Functional programming, there are four types of identifiers.
1. Local identifiers
2. Global identifiers
3. Non-local identifiers
4. Built-in identifiers

```python
#Local Identifiers :-
def display(name):
    #local variable
    age = 22
    print(f"{name} has age {age}")

display("shantanu")
```

```python
#scope of local identifiers/variables
def display(name:str) -> None:
    #local variable
    age:int = 22
    print(f"{name} has age {age}") # name is treated as local variable
    print(locals()) # this function is used to show the local variables
    print(len(locals())) # it will return count of all local variables

try:
    display("shantanu")
    print(age) # if we try to access it outside the function it will throw NameError: 'age' is not define
except Exception as e:
    print(f"{e}.")
```

```python
# GLOBAL Variable
from typing import Final

COUNTRY:Final[str] = 'UK'

def show_country(name:str) -> None:
    age:int = 22
    print(f"{name} belongs to {COUNTRY} and has age {age}")

show_country('Jay')
print(COUNTRY)
```

```python
#what if we have same local and global variable
from typing import Final
NUM:Final[int] = 10 #global variable

def display() -> None:
    num:int = 20 # local variable
    print("Inside:",num)

display()
print('Outside:',NUM)
```

#### GLOBAL Keyword and UnboundLocalError

```python
# https://claude.ai/share/7f516700-25af-4e7a-ba82-d643787a7690
from typing import Final

NUM:Final[int] = 10 #global variable PEP 8: constants must be written in UPPER_CASE

def display() -> None:
    """
    Demonstrate the use of global keyword and variable scoping in Python.

    - Global variable NUM is accessed and modified inside the function.
    - A separate local variable 'num' is used to avoid annotation conflict.
    """
    global NUM # global keyword allows to change value of NUM variable inside the function

    num:int = NUM + 5 # use of global variable value inside local variable
    print(f"After using GLOBAL variable NUM, local num value : ", {num})

    num = 20 # assign new value to the local variable
    print('Local num variable new value: ',num)

    NUM = 99 # modify GLOBAL NUM value
    print(f"new value of GLOBAL NUM variable inside function : ", {NUM})

#function call
display()

print('Final value GLOBAL variable outside the function: ',NUM)
```

```python
NUM = 10

def only_read() -> None:
    # could be read inside without global
    print(NUM)

def modify() -> None:
    global NUM  # only use when we need to modify its value
    NUM = 99
    print(NUM)

only_read()
modify()
```

### `return` statements in python

- Need of return statement
- How to return values from function ?
- Some important points to remember
- we can return any type of python object
- return statement must be last statement of function, because nothing will execute after it
- we can use return keyword to stop execution of function, just like break in loops

```python
#Need of return statement
# this is a simple function that calculate the simple interest.
# but what is we need total amount payable:- Interest + Principal
# si is local scope variable and we can't access it outside the function
# that's why we return the value, so that we use it other ways


def simple_interest(principal_amount:int,number_of_years:int ,rate_of_interest:float) -> float:
    si = principal_amount * number_of_years * rate_of_interest
    # print("Simple interest is:",si)
    return si

P = 10_000
N = 9
R = 0.0925
print(f"Simple Interest for {N} years is : {simple_interest(P,N,R):,.2f} USD.")
print(f"Total payable amount is: {simple_interest(P,N,R) + P:,.2f} USD")
# total = P + si
```

```python
# You can return multiple values in the shape of tuple object.
# But in reality it is returning just one object(tuple) at a time

def calci(a:int,b:int)-> tuple[int, int] :
    add:int = a + b
    sub:int = a - b
    return add, sub

print(calci(50,20))
print(f"The Addition value is: {calci(50,20)[0]} and Subtraction value is: {calci(50,20)[1]}.")
```

```python
#we can only use return keyword to stop execution of function

def calci(a:int,b:int)-> tuple[int, int] | None:
    if a < 0:
        return None
    add = a + b
    sub = a - b
    return add , sub

#n1,n2 = calci(-10,20)
# it throws this error: TypeError: cannot unpack non-iterable NoneType object
# because in place of calci(-10,20) now its None.becuase we enter -10 value, and return keyword stops the program. and None can't go inside 2 variables, that's why we got this error

#but if we do like this, it will be ok
print(calci(50,20))

#but if we do like this, it will be return None
print(calci(-50,20))

# print("addition is:",n1)
# print("subtraction is:",n2)
```
