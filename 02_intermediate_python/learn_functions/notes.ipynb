#### Scope in programming

In programming, the term "scope" refers to the region of the code where an identifier can be accessed.

#### What is an identifier ?

an identifier is a name using which we can identify a particular object. Objects can be variable, function name, class name etc.

In Functional programming, there are four types of identifiers.
1. Local identifiers
2. Global identifiers
3. Non-local identifiers
4. Built-in identifiers

````python
#Local Identifiers :-
def display(name):
    #local variable
    age = 22
    print(f"{name} has age {age}")

display("shantanu")
````

````python
#scope of local identifiers/variables
def display(name:str) -> None:
    #local variable
    age:int = 22
    print(f"{name} has age {age}") # name is treated as local variable
    print(locals()) # this function is used to show the local variables
    print(len(locals())) # it will return count of all local variables

try:
    display("shantanu")
    print(age) # if we try to access it outside the function it will throw NameError: 'age' is not define
except Exception as e:
    print(f"{e}.")
````

#### GLOBAL Keyword and UnboundLocalError

````python
# https://claude.ai/share/7f516700-25af-4e7a-ba82-d643787a7690
from typing import Final

NUM:Final[int] = 10 #global variable PEP 8: constants must be written in UPPER_CASE

def display() -> None:
    """
    Demonstrate the use of global keyword and variable scoping in Python.

    - Global variable NUM is accessed and modified inside the function.
    - A separate local variable 'num' is used to avoid annotation conflict.
    """
    global NUM # global keyword allows to change value of NUM variable inside the function

    num:int = NUM + 5 # use of global variable value inside local variable
    print(f"After using GLOBAL variable NUM, local num value : ", {num})

    num = 20 # assign new value to the local variable
    print('Local num variable new value: ',num)

    NUM = 99 # modify GLOBAL NUM value
    print(f"new value of GLOBAL NUM variable inside function : ", {NUM})

#function call
display()

print('Final value GLOBAL variable outside the function: ',NUM)
````

````python
NUM = 10

def only_read() -> None:
    # could be read inside without global
    print(NUM)

def modify() -> None:
    global NUM  # only use when we need to modify its value
    NUM = 99
    print(NUM)

only_read()
modify()
````

### `return` statements in python

- Need of return statement
- How to return values from function ?
- Some important points to remember
- we can return any type of python object
- return statement must be last statement of function, because nothing will execute after it
- we can use return keyword to stop execution of function, just like break in loops

````python
#Need of return statement
# this is a simple function that calculate the simple interest.
# but what is we need total amount payable:- Interest + Principal
# si is local scope variable and we can't access it outside the function
# that's why we return the value, so that we use it other ways


def simple_interest(principal_amount:int,number_of_years:int ,rate_of_interest:float) -> float:
    si = principal_amount * number_of_years * rate_of_interest
    # print("Simple interest is:",si)
    return si

P = 10_000
N = 9
R = 0.0925
print(f"Simple Interest for {N} years is : {simple_interest(P,N,R):,.2f} USD.")
print(f"Total payable amount is: {simple_interest(P,N,R) + P:,.2f} USD")
# total = P + si
````

````python
# You can return multiple values in the shape of tuple object.
# But in reality it is returning just one object(tuple) at a time

def calci(a:int,b:int)-> tuple[int, int] :
    add:int = a + b
    sub:int = a - b
    return add, sub

print(calci(50,20))
print(f"The Addition value is: {calci(50,20)[0]} and Subtraction value is: {calci(50,20)[1]}.")
````

### Positional and Keyword Arguments

Parameters :-

- Variables for holding actual values
- Present in function definition

Arguments :-
- Values provided at function call

#### Types of Arguments

1. Positional arguments
2. Keyword arguments
3. Default arguments
4. Variable length positional arguments
5. Variable length keyword arguments

#### Exact order to Follow, if mixing multiple types of arguments:

positional arguments → variable length positional arguments → keyword arguments → variable length keyword arguments → default arguments

#### Examples

````python
# positional arguments

"""
The order in which you pass the values during the function call,  should match the order in which the parameters are defined in the function.
we have to follow the order of the parameters, if we change the order of arguments, result will be wrong.
Number of parameters = Number of arguments
"""
def display(name:str, age:int) -> None:
    print(f"name is {name} and age is {age}")

display('sara', 23)
````

````python
# Keyword arguments

"""
in positional arguments we have to follow the position of the parameters defined in the function definition
for example:- function definition is hidden somewhere else, we dont know their positions, so how we handle them that's why we use Keyword Arguments
order of positional arguments have no importance
but we have to keep number of parameters and arguments same
"""

def display(name:str, age:int) -> None:
    print(f"name is {name} and age is {age}")

display(age=52, name='sara')
display('sara', age=15) # mixing positional and keyword arguments following above heirarchy
````

#### Mixing Variable Length (Positional and Keyword) Arguments(*Args **KWArgs) - Ordering Arguments

````python
# mixing variable length keyword and positional arguments
"""
1. Standard arguments
2. *args arguments
3. **kwargs arguments
"""
# we cannot use *args after **kwargs
# SyntaxError: arguments cannot follow var-keyword argument, if we do so

# correct code
def addition(*num:int, **nums:int) -> int:
    return sum(num) + sum(nums.values())

print(addition(100,200,300, n1 = 10,n2 = 20,n3 = 30))
````

#### Examples of *Args and **Kwargs

````python
# Unpacking With the Asterisk Operators: * & **

"""
The single and double asterisk unpacking operators were introduced in Python 2. As of the 3.5 release, they have become even more powerful, thanks to PEP 448. In short, the unpacking operators are operators that unpack the values from iterable objects in Python. The single asterisk operator * can be used on any iterable that Python provides, while the double asterisk operator ** can only be used on dictionaries.
"""
my_list = [1, 2, 3]
print(*my_list)
````

````python
def my_sum(*args:int) -> int:
    result = 0
    for x in args:
        result += x
    return result

list1 = [1, 2, 3]
list2 = [4, 5]
list3 = [6, 7, 8, 9]

print(my_sum(*list1, *list2, *list3))
````

There are other **convenient uses** of the **unpacking operator**. For example, say you need to **split** a list into three different parts. The output should show the **first value**, the **last value**, and **all the values in between**. With the unpacking operator, you can do this in just one line of code:

````python
my_list:list[int] = [1, 2, 3, 4, 5, 6]

a, *b, c = my_list

print(a)
print(b)
print(c)
````

Another interesting thing you can do with the unpacking operator * is to split the items of any iterable object. This could be very useful if you need to merge two lists, for instance:

````python
my_first_list = [1, 2, 3]
my_second_list = [4, 5, 6]
my_merged_list = [*my_first_list, *my_second_list]
# my_merged_list = my_first_list + my_second_list # we can do by this

print(my_merged_list)
````

````python
def describe_pet(name:str, animal_type:str, ** details:Any) -> None:
    print(f"Pet name: {name}")
    print(f"Animal type: {animal_type}")

    if details:
        print("Details:")
    for key, value in details.items():
        print(f" {key}: {value}")

# Example usage:
describe_pet("Whiskers", "cat", color="gray", age=3, favorite_food="tuna")
````

````python
def make_sandwich(*ingredients:str, ** options:Any) -> None:
    print("Make a sandwich with:")
    for ingredient in ingredients:
        print(f"- {ingredient}")

    print("Option:")
    for key, value in options.items ():
        print(f"- {key}: {value}")

# Example usage:
make_sandwich("turkey", "lettuce", "tomato", sauce="mayo", toasted=True)
````
