## creating class and objects

so, in python we can define our own Classes(data structures) and then create objects knows as instances of that class.

Now Let's create our own class.







```python
class Dog:
    """Represent a dog with a name and age.

    Demonstrates a basic class definition in Python — a blueprint
    for creating Dog objects (instances) with shared attributes
    and behaviour.

    Attributes:
        name (str): The name of the dog.
        age (int): The age of the dog in years.

    Methods:
        bark(): Print a barking message for this dog.

    Example:
        >>> dog = Dog("Rex", 3)
        >>> dog.bark()
        Rex is barking.
    """

    def __init__(self, name: str, age: int) -> None:
        """Initialise a Dog with a name and age.

        Args:
            name (str): The name of the dog.
            age (int): The age of the dog in years.
        """
        self.name: str = name  # instance attribute(variable)
        self.age: int = age  # instance attribute(variable)

    def bark(self) -> None:  # instance method(function)
        """Print a barking message for this dog.

        Example:
            >>> dog = Dog("Rex", 3)
            >>> dog.bark()
            Rex is barking.
        """
        print(f"{self.name} is barking.")
```

now let's create object(instance) of this class.

```python
my_dog: Dog = Dog("Buddy", 3)
my_dog.bark()

john_dog: Dog = Dog("Tommy", 5)
john_dog.bark()

mahira_dog: Dog = Dog("Sheena", 8)
mahira_dog.bark()
```

### Combining Objects

let's assume every dog must have its owner. and owner have name, address, age, etc.

```python
class Owner:
    """Represent a dog owner with personal details.

    Attributes:
        name (str): The full name of the owner.
        address (str): The home address of the owner.
        age (int): The age of the owner in years.

    Example:
        >>> owner = Owner("John", "123 Main St", 30)
        >>> owner.name
        'John'
    """

    def __init__(self, name: str, address: str, age: int) -> None:
        """Initialise an Owner with name, address, and age.

        Args:
            name (str): The full name of the owner.
            address (str): The home address of the owner.
            age (int): The age of the owner in years.
        """
        self.name: str = name
        self.address: str = address
        self.age: int = age
```

```python
class Dog:
    """Represent a dog with a name, age, and an owner.

    Demonstrates object composition in Python — a Dog *has an* Owner,
    meaning an Owner object is embedded as an attribute rather than
    using plain strings to store owner details.

    Attributes:
        name (str): The name of the dog.
        age (int): The age of the dog in years.
        owner (Owner): The Owner instance associated with this dog.
            See :class:`Owner` for available attributes such as
            ``owner.name``, ``owner.address``, and ``owner.age``.

    Methods:
        bark(): Print a barking message for this dog.

    Example:
        >>> owner = Owner("John", "123 Main St", 30)
        >>> dog = Dog("Rex", 3, owner)
        >>> dog.bark()
        Rex is barking.
        >>> dog.owner.name
        'John'
    """

    def __init__(self, name: str, age: int, owner: Owner) -> None:
        """Initialise a Dog with a name, age, and owner.

        Args:
            name (str): The name of the dog.
            age (int): The age of the dog in years.
            owner (Owner): The Owner instance responsible for this dog.
        """
        self.name: str = name  # instance attribute(variable)
        self.age: int = age  # instance attribute(variable)
        self.owner: Owner = owner  # composition — Dog has an Owner

    def bark(self) -> None:  # instance method(function)
        """Print a barking message for this dog.

        Example:
            >>> owner = Owner("John", "123 Main St", 30)
            >>> dog = Dog("Rex", 3, owner)
            >>> dog.bark()
            Rex is barking.
        """
        print(f"{self.name} is barking.")
```

```python
my_details: Owner = Owner("John Smith", "123 Main St", 30)
my_dog: Dog = Dog("Buddy", 3, my_details)
my_dog.bark()
print(f"My owner name is: {my_dog.owner.name}")

mahira_details: Owner = Owner("Mahira", "123 Main St", 30)
mahira_dog: Dog = Dog("Sheena", 8, mahira_details)
mahira_dog.bark()
print(f"My owner name is: {mahira_dog.owner.name}")
```

### creating classes with dataclasses - built-in module

```python
from dataclasses import dataclass


@dataclass
class Person:
    name: str
    age: int
    address: str

    def greet(self) -> str:
        return f"Hello, my name is {self.name} and I am {self.age} years old."
```

## Public, Protected, Private attributes

| Attribute Type | Prefix                 | Access Level                                     | Use Case                                         |
| -------------- | ---------------------- | ------------------------------------------------ | ------------------------------------------------ |
| Public         | None                   | Accessible everywhere                            | Part of the intended API                         |
| Protected      | _ (single underscore)  | Accessible but discouraged outside class         | Internal implementation, but okay for subclasses |
| Private        | __ (double underscore) | Name-mangled; strongly discouraged outside class | Truly internal logic you want to hide            |

#### Public attributes



Public attributes are accessible from anywhere—both inside and outside the class. They're named normally without any special prefix. This is the default in Python.

```python
class User:
    def __init__(self, name: str, age: int, address: str) -> None:
        self.name: str = name
        self.age: int = age
        self.address: str = address
```

```python
john: User = User("John", 30, "123 Main St")
print(john.name)
print(john.age)
print(john.address)
```

#### Protected attributes

**Protected attributes** are intended for **internal use within the class and its subclasses**, but are still technically accessible from outside. They use a single underscore prefix `_.attribute_name` by convention.

The single underscore(_) is a naming convention, not enforced by Python. It's a signal to other developers: "This is internal—use the public interface instead."

```python
class User:
    def __init__(self, name: str, age: int, address: str) -> None:
        self._name: str = name
        self._age: int = age
        self._address: str = address

    # getter method
    def get_name(self) -> str:
        return self._name

    # getter method
    def get_age(self) -> int:
        return self._age

    # getter method
    def get_address(self) -> str:
        return self._address
```

```python
user: User = User("John", 30, "123 Main St")
print(user.get_name())
print(user.get_age())
print(user.get_address())
```

#### Private attributes

**Private attributes** use **double underscore** prefix (__) and trigger Python's name mangling, making them harder to access from **outside the class**. They're **intended to be truly internal**.

```python
class BankAccount:
    def __init__(self, balance: int) -> None:
        self.__balance: int = balance  # private attribute

    # getter method
    def get_balance(self) -> int:
        return self.__balance
```

```python
account: BankAccount = BankAccount(1000)
print(account.get_balance())  # Correct way: 1000
```

#### Getters and Setters methods

The best Python practice is to use the `@property` decorator(AKA getter method) instead of `explicit` **getter and setter methods**. This allows clean dot notation while **maintaining encapsulation**.

```python
class User:
    """Represent a user with validated, encapsulated attributes.

    Demonstrates the Pythonic approach to encapsulation using
    ``@property`` decorators instead of explicit getter/setter methods.
    Properties let callers use clean dot-notation (``user.name``) while
    still enforcing validation on every assignment.

    Attributes:
        name (str): The full name of the user.
        age (int): The age of the user in years (must be >= 0).
        address (str): The home address of the user.

    Example:
        >>> user = User("Alice", 30, "123 Main St")
        >>> user.name
        'Alice'
        >>> user.age = 31
        >>> user.age
        31
        >>> user.age = -1
        Traceback (most recent call last):
            ...
        ValueError: Age must be a non-negative integer
    """

    def __init__(self, name: str, age: int, address: str) -> None:
        """Initialise a User with name, age, and address.

        Args:
            name (str): The full name of the user.
            age (int): The age of the user in years.
            address (str): The home address of the user.
        """
        self._name: str = name
        self._age: int = age
        self._address: str = address

    def __repr__(self) -> str:
        """Return an unambiguous string representation of this User."""
        return f"User(name={self._name!r}, age={self._age!r}, address={self._address!r})"

    @property  # it is a decorator that allows us to access the attribute as a property. AKA getter method
    def name(self) -> str:  # method name should be same as attribute name
        """The full name of the user."""
        return self._name

    @name.setter  # it is a decorator that allows us to set the attribute as a property. AKA setter method
    def name(self, value: str) -> None:  # method name should be same as attribute name
        if not isinstance(value, str) or not value.strip():
            raise ValueError("Name must be a non-empty string")
        self._name = value

    @property  # it is a decorator that allows us to access the attribute as a property. AKA getter method
    def age(self) -> int:  # method name should be same as attribute name
        """The age of the user in years."""
        return self._age

    @age.setter  # it is a decorator that allows us to set the attribute as a property. AKA setter method
    def age(self, value: int) -> None:  # method name should be same as attribute name
        if not isinstance(value, int) or value < 0:
            raise ValueError("Age must be a non-negative integer")
        self._age = value

    @property  # it is a decorator that allows us to access the attribute as a property. AKA getter method
    def address(self) -> str:  # method name should be same as attribute name
        """The home address of the user."""
        return self._address

    @address.setter  # it is a decorator that allows us to set the attribute as a property. AKA setter method
    def address(self, value: str) -> None:  # method name should be same as attribute name
        if not isinstance(value, str) or not value.strip():
            raise ValueError("Address must be a non-empty string")
        self._address = value
```

```python
user: User = User("Alice", 30, "123 Main St")

# __repr__ gives an unambiguous snapshot of the object
print(repr(user))
print("*" * 35)

# Access properties with clean dot-notation
print(user.name)     # Alice
print(user.age)      # 30
print(user.address)  # 123 Main St
print("*" * 35)

# Setters validate on assignment — no explicit method call needed
user.name = "Bob"
user.age = 35
user.address = "456 Oak Ave"

print(repr(user))
print("*" * 35)

# Invalid values raise errors at the point of assignment
try:
    user.age = -5
except ValueError as e:
    print(e)  # Age must be a non-negative integer

try:
    user.name = ""
except ValueError as e:
    print(e)  # Name must be a non-empty string
```
