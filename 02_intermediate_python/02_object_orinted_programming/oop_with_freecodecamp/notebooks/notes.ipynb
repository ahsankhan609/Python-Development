## OOP Fundamentals

so, in python we can define our own Classes(data structures) and then create objects knows as instances of that class.

Now Let's create our own class.







````python
class Dog:
    """Represent a dog with a name and age.

    Demonstrates a basic class definition in Python — a blueprint
    for creating Dog objects (instances) with shared attributes
    and behaviour.

    Attributes:
        name (str): The name of the dog.
        age (int): The age of the dog in years.

    Methods:
        bark(): Print a barking message for this dog.

    Example:
        >>> dog = Dog("Rex", 3)
        >>> dog.bark()
        Rex is barking.
    """

    def __init__(self, name: str, age: int) -> None:
        """Initialise a Dog with a name and age.

        Args:
            name (str): The name of the dog.
            age (int): The age of the dog in years.
        """
        self.name: str = name  # instance attribute(variable)
        self.age: int = age  # instance attribute(variable)

    def bark(self) -> None:  # instance method(function)
        """Print a barking message for this dog.

        Example:
            >>> dog = Dog("Rex", 3)
            >>> dog.bark()
            Rex is barking.
        """
        print(f"{self.name} is barking.")
````

now let's create object(instance) of this class.

````python
my_dog: Dog = Dog("Buddy", 3)
my_dog.bark()

john_dog: Dog = Dog("Tommy", 5)
john_dog.bark()

mahira_dog: Dog = Dog("Sheena", 8)
mahira_dog.bark()
````

### Combining Objects

let's assume every dog must have its owner. and owner have name, address, age, etc.

````python
class Owner:
    """Represent a dog owner with personal details.

    Attributes:
        name (str): The full name of the owner.
        address (str): The home address of the owner.
        age (int): The age of the owner in years.

    Example:
        >>> owner = Owner("John", "123 Main St", 30)
        >>> owner.name
        'John'
    """

    def __init__(self, name: str, address: str, age: int) -> None:
        """Initialise an Owner with name, address, and age.

        Args:
            name (str): The full name of the owner.
            address (str): The home address of the owner.
            age (int): The age of the owner in years.
        """
        self.name: str = name
        self.address: str = address
        self.age: int = age
````

````python
class Dog:
    """Represent a dog with a name, age, and an owner.

    Demonstrates object composition in Python — a Dog *has an* Owner,
    meaning an Owner object is embedded as an attribute rather than
    using plain strings to store owner details.

    Attributes:
        name (str): The name of the dog.
        age (int): The age of the dog in years.
        owner (Owner): The Owner instance associated with this dog.
            See :class:`Owner` for available attributes such as
            ``owner.name``, ``owner.address``, and ``owner.age``.

    Methods:
        bark(): Print a barking message for this dog.

    Example:
        >>> owner = Owner("John", "123 Main St", 30)
        >>> dog = Dog("Rex", 3, owner)
        >>> dog.bark()
        Rex is barking.
        >>> dog.owner.name
        'John'
    """

    def __init__(self, name: str, age: int, owner: Owner) -> None:
        """Initialise a Dog with a name, age, and owner.

        Args:
            name (str): The name of the dog.
            age (int): The age of the dog in years.
            owner (Owner): The Owner instance responsible for this dog.
        """
        self.name: str = name  # instance attribute(variable)
        self.age: int = age  # instance attribute(variable)
        self.owner: Owner = owner  # composition — Dog has an Owner

    def bark(self) -> None:  # instance method(function)
        """Print a barking message for this dog.

        Example:
            >>> owner = Owner("John", "123 Main St", 30)
            >>> dog = Dog("Rex", 3, owner)
            >>> dog.bark()
            Rex is barking.
        """
        print(f"{self.name} is barking.")
````

````python
my_details: Owner = Owner("John Smith", "123 Main St", 30)
my_dog: Dog = Dog("Buddy", 3, my_details)
my_dog.bark()
print(f"My owner name is: {my_dog.owner.name}")

mahira_details: Owner = Owner("Mahira", "123 Main St", 30)
mahira_dog: Dog = Dog("Sheena", 8, mahira_details)
mahira_dog.bark()
print(f"My owner name is: {mahira_dog.owner.name}")
````

### Public, Protected, Private attributes

| Attribute Type | Prefix                 | Access Level                                     | Use Case                                         |
| -------------- | ---------------------- | ------------------------------------------------ | ------------------------------------------------ |
| Public         | None                   | Accessible everywhere                            | Part of the intended API                         |
| Protected      | _ (single underscore)  | Accessible but discouraged outside class         | Internal implementation, but okay for subclasses |
| Private        | __ (double underscore) | Name-mangled; strongly discouraged outside class | Truly internal logic you want to hide            |

#### Public attributes



Public attributes are accessible from anywhere—both inside and outside the class. They're named normally without any special prefix. This is the default in Python.

````python
class User:
    def __init__(self, name: str, age: int, address: str) -> None:
        self.name: str = name
        self.age: int = age
        self.address: str = address
````

````python
john: User = User("John", 30, "123 Main St")
print(john.name)
print(john.age)
print(john.address)
````

#### Protected attributes

**Protected attributes** are intended for **internal use within the class and its subclasses**, but are still technically accessible from outside. They use a single underscore prefix `_.attribute_name` by convention.

The single underscore(_) is a naming convention, not enforced by Python. It's a signal to other developers: "This is internal—use the public interface instead."

````python
class User:
    def __init__(self, name: str, age: int, address: str) -> None:
        self._name: str = name
        self._age: int = age
        self._address: str = address

    # getter method
    def get_name(self) -> str:
        return self._name

    # getter method
    def get_age(self) -> int:
        return self._age

    # getter method
    def get_address(self) -> str:
        return self._address
````

````python
user: User = User("John", 30, "123 Main St")
print(user.get_name())
print(user.get_age())
print(user.get_address())
````

#### Private attributes

**Private attributes** use **double underscore** prefix (__) and trigger Python's name mangling, making them harder to access from **outside the class**. They're **intended to be truly internal**.

````python
class BankAccount:
    def __init__(self, balance: int) -> None:
        self.__balance: int = balance  # private attribute

    # getter method
    def get_balance(self) -> int:
        return self.__balance
````

````python
account: BankAccount = BankAccount(1000)
print(account.get_balance())  # Correct way: 1000
````

#### Getters and Setters methods - `@property` decorator - to access attributes in instance - '@attribute_name.setter' decorator to set the attribute

The best Python practice is to use the `@property` decorator(AKA getter method) instead of `explicit` **getter and setter methods**. This allows clean dot notation while **maintaining encapsulation**.

````python
class User:
    """Represent a user with validated, encapsulated attributes.

    Demonstrates the Pythonic approach to encapsulation using
    ``@property`` decorators instead of explicit getter/setter methods.
    Properties let callers use clean dot-notation (``user.name``) while
    still enforcing validation on every assignment.

    Attributes:
        name (str): The full name of the user.
        age (int): The age of the user in years (must be >= 0).
        address (str): The home address of the user.

    Example:
        >>> user = User("Alice", 30, "123 Main St")
        >>> user.name
        'Alice'
        >>> user.age = 31
        >>> user.age
        31
        >>> user.age = -1
        Traceback (most recent call last):
            ...
        ValueError: Age must be a non-negative integer
    """

    def __init__(self, name: str, age: int, address: str) -> None:
        """Initialise a User with name, age, and address.

        Args:
            name (str): The full name of the user.
            age (int): The age of the user in years.
            address (str): The home address of the user.
        """
        self._name: str = name
        self._age: int = age
        self._address: str = address

    def __repr__(self) -> str:
        """Return an unambiguous string representation of this User."""
        return f"User(name={self._name!r}, age={self._age!r}, address={self._address!r})"

    @property  # it is a decorator that allows us to access the attribute as a property. AKA getter method
    def name(self) -> str:  # method name should be same as attribute name
        """The full name of the user."""
        return self._name

    @name.setter  # it is a decorator that allows us to set the attribute as a property. AKA setter method
    def name(self, new_name: str) -> None:  # method name should be same as attribute name
        if not isinstance(new_name, str) or not new_name.strip():
            raise ValueError("Name must be a non-empty string")
        self._name = new_name

    @property  # it is a decorator that allows us to access the attribute as a property. AKA getter method
    def age(self) -> int:  # method name should be same as attribute name
        """The age of the user in years."""
        return self._age

    @age.setter  # it is a decorator that allows us to set the attribute as a property. AKA setter method
    def age(self, new_age: int) -> None:  # method name should be same as attribute name
        if not isinstance(new_age, int) or new_age < 0:
            raise ValueError("Age must be a non-negative integer")
        self._age = new_age

    @property  # it is a decorator that allows us to access the attribute as a property. AKA getter method
    def address(self) -> str:  # method name should be same as attribute name
        """The home address of the user."""
        return self._address

    @address.setter  # it is a decorator that allows us to set the attribute as a property. AKA setter method
    # method name should be same as attribute name
    def address(self, new_address: str) -> None:
        if not isinstance(new_address, str) or not new_address.strip():
            raise ValueError("Address must be a non-empty string")
        self._address = new_address
````

with this property decorator, we don't need to call the method to access the attribute. we can access the attribute directly. it is easy to use and more readable.

we can make any attribute private or protected, when need easily.

**Why This Is Best Practice**:

**Cleaner syntax:** `user.name` instead of `user.get_name()`—it reads like a normal attribute.

**Maintainability:** You can add validation or logic without breaking code that uses the class.

**Future-proof:** If you later need complex logic, properties handle it gracefully without requiring API changes.

**Pythonic:** This follows PEP 8 and Python conventions; explicit getter/setter methods are considered un-Pythonic.

**Alternative: Read-Only Properties** - Helps in Encapsulation

If you want a property that **can't be modified**, `omit the setter`:
it is useful in Encapsulation.

```python
@property
def name(self) -> str:
    return self._name


# No setter → read-only. user.name = "Bob" raises AttributeError
```

````python
user: User = User("Alice", 30, "123 Main St")

# __repr__ gives an unambiguous snapshot of the object
print(repr(user))
print("*" * 35)

# Access properties with clean dot-notation
print(user.name)     # Alice
print(user.age)      # 30
print(user.address)  # 123 Main St
print("*" * 35)

# Setters validate on assignment — no explicit method call needed
user.name = "Bob"
user.age = 35
user.address = "456 Oak Ave"

print(repr(user))
print("*" * 35)

# Invalid values raise errors at the point of assignment
try:
    user.age = -5
except ValueError as e:
    print(e)  # Age must be a non-negative integer

try:
    user.name = ""
except ValueError as e:
    print(e)  # Name must be a non-empty string
````

### Static(class) Attributes(variable) - For Python Class

previously we know that **each object(instance) of the class** has its **own attributes(variables) (called as instance attributes(variables))**.

but sometimes we need to store some attributes(variables) that are **common/shared to all objects(instances) of the class**. for this we use **Static Attributes.**

A **Static Attribute** (sometimes called a class attribute) is an attribute, that belongs to the **class itself**, not to **any specific instance** of the class.

it also means that only one copy of class attribute is created in the memory, regardless of how many instances of the class inctances are created.

class attributes are accessed directly from the class, not through the instance of the class.      

we use `@classmethod` decorator to access static(class) attributes(variables).

we use `@staticmethod` to access utility functions(regular functions, directly related to the class).


````python
class Car:
    # Class attributes
    COMPANY_NAME: str = "Toyota"
    COMPANY_LOCATION: str = "Japan"

    def __init__(self, name: str,
                 manufactured_year: int,
                 color: str,
                 wheels: int = 4) -> None:
        self._name: str = name
        self._manufactured_year: int = manufactured_year
        self._color: str = color
        self._wheels: int = wheels

    # Instance properties with validation
    @property  # getter method for instance attribute
    def name(self) -> str:
        return self._name

    @name.setter  # setter method for instance attribute
    def name(self, value: str) -> None:
        if not isinstance(value, str) or not value.strip():
            raise ValueError("Name must be a non-empty string")
        self._name = value

    @property
    def manufactured_year(self) -> int:
        return self._manufactured_year

    @manufactured_year.setter
    def manufactured_year(self, value: int) -> None:
        if not isinstance(value, int) or value < 1886:
            raise ValueError("Year must be >= 1886 (first car)")
        self._manufactured_year = value

    @property
    def color(self) -> str:
        return self._color

    @color.setter
    def color(self, value: str) -> None:
        if not isinstance(value, str) or not value.strip():
            raise ValueError("Color must be a non-empty string")
        self._color = value

    @property
    def wheels(self) -> int:
        return self._wheels

    @wheels.setter
    def wheels(self, value: int) -> None:
        if not isinstance(value, int) or value <= 0:
            raise ValueError("Wheels must be positive")
        self._wheels = value

    # Magic methods for string representation - user friendly representation
    def __str__(self) -> str:
        """User-friendly representation"""
        return f"Car Name: {self.name}, Manufactured Year: {self.manufactured_year}, Color: {self.color}, Wheels: {self.wheels}"

    def __repr__(self) -> str:  # developer friendly representation
        """Developer-friendly representation"""
        return f"Car(name={self.name!r}, manufactured_year={self.manufactured_year}, color={self.color!r}, wheels={self.wheels})"

    @classmethod  # class method decorator to access class attributes
    def get_company_details(cls) -> str:
        """Class method for company information"""
        return f"Company Name: {cls.COMPANY_NAME}, Company Location: {cls.COMPANY_LOCATION}"
````

**Key Improvements**

| Aspect                  | Your Code                                        | Improved Code                          |
| ----------------------- | ------------------------------------------------ | -------------------------------------- |
| String representation   | get_car_details(), get_company_details() methods | __str__() and __repr__() magic methods |
| Encapsulation           | Public attributes (no protection)                | Protected attributes with @property    |
| Validation              | None                                             | Input validation in setters            |
| Class attributes        | self.__class__.COMPANY_NAME                      | cls.COMPANY_NAME in classmethod        |
| Accessing instance data | Direct attribute access                          | Property-based access with control     |

**When to Use Each**

* __str__() — for print(obj) and str(obj) — user-friendly
* __repr__() — for repr(obj) and debugger display — developer-friendly
* @classmethod — when a method needs class data (like company info)
* `@property` — when you need to control attribute access with validation

````python
toyota = Car("Toyota", 2020, "Red")


# Use magic methods (clean printing)
print(toyota)  # Car Name: Toyota, Manufactured Year: 2020, Color: Red, Wheels: 4
# Car(name='Toyota', manufactured_year=2020, color='Red', wheels=4)
print(repr(toyota))


# Access properties (like attributes)
print(toyota.name)  # Toyota
toyota.color = "Blue"  # Sets via property


# Class method for company info
# Company Name: Toyota, Company Location: Japan
print(Car.get_company_details())


# Validation works
toyota.manufactured_year = 1900  # ValueError: Year must be >= 1886
````

**STATIC METHOD**:

* No `self` or `cls` parameter
* Cannot access instance or class variables
* Acts like a `regular function` but logically belongs to the class
* Used for utility/helper functions related to the class topic

````python
from datetime import datetime


class Car:
    # Class attributes
    COMPANY_NAME: str = "Toyota"
    COMPANY_LOCATION: str = "Japan"

    def __init__(self, name: str,
                 manufactured_year: int,
                 color: str,
                 wheels: int = 4) -> None:
        self._name: str = name
        self._manufactured_year: int = manufactured_year
        self._color: str = color
        self._wheels: int = wheels

    # Instance properties with validation

    @property
    def name(self) -> str:
        return self._name

    @name.setter
    def name(self, value: str) -> None:
        if not isinstance(value, str) or not value.strip():
            raise ValueError("Name must be a non-empty string")
        self._name = value

    @property
    def manufactured_year(self) -> int:
        return self._manufactured_year

    @manufactured_year.setter
    def manufactured_year(self, value: int) -> None:
        if not isinstance(value, int) or value < 1886:
            raise ValueError("Year must be >= 1886 (first car)")
        self._manufactured_year = value

    @property
    def color(self) -> str:
        return self._color

    @color.setter
    def color(self, value: str) -> None:
        if not isinstance(value, str) or not value.strip():
            raise ValueError("Color must be a non-empty string")
        self._color = value

    @property
    def wheels(self) -> int:
        return self._wheels

    @wheels.setter
    def wheels(self, value: int) -> None:
        if not isinstance(value, int) or value <= 0:
            raise ValueError("Wheels must be positive")
        self._wheels = value

    # Magic methods for string representation

    def __str__(self) -> str:
        """User-friendly representation"""
        return f"Car Name: {self.name}, Manufactured Year: {self.manufactured_year}, Color: {self.color}, Wheels: {self.wheels}"

    def __repr__(self) -> str:
        """Developer-friendly representation"""
        return f"Car(name={self.name!r}, manufactured_year={self.manufactured_year}, color={self.color!r}, wheels={self.wheels})"

    @classmethod
    def get_company_details(cls) -> str:
        """Class method for company information"""
        return f"Company Name: {cls.COMPANY_NAME}, Company Location: {cls.COMPANY_LOCATION}"

    # Static methods - utility functions not dependent on instance or class data

    @staticmethod
    def calculate_car_age(manufactured_year: int) -> int:
        """
        Calculate the age of a car based on its manufacturing year.
        Does not need self or cls since it's a utility function.
        """
        current_year: int = datetime.now().year
        age: int = current_year - manufactured_year

        if age < 0:
            raise ValueError(f"Manufacturing year cannot be in the future")
        return age

    @staticmethod
    def is_vintage_car(manufactured_year: int) -> bool:
        """
        Check if a car is considered vintage (25+ years old).
        Independent utility function related to car classification.
        """
        age: int = Car.calculate_car_age(manufactured_year)
        return age >= 25

    @staticmethod
    def is_valid_color(color: str) -> bool:
        """
        Validate if a color is in the list of accepted car colors.
        Pure utility function with no instance/class dependency.
        """
        valid_colors: list[str] = [
            "Red", "Blue", "Black", "White", "Silver",
            "Gray", "Green", "Yellow", "Orange", "Brown",  "Purple"
        ]
        return color.strip().capitalize() in valid_colors
````

````python
# Create car instance
toyota = Car("Camry", 2020, "Red")
print(toyota)


# Instance method - get company details
print(Car.get_company_details())


# Static method 1: Calculate car age
age = Car.calculate_car_age(2020)
print(f"Car age: {age} years")  # Car age: 6 years


# Static method 2: Check if vintage
is_vintage = Car.is_vintage_car(1995)

# Is 1995 car vintage? True
print(f"Is 1995 car vintage? {is_vintage}")


# Static method 3: Validate color
valid = Car.is_valid_color("red")
print(f"Is 'red' a valid color? {valid}")  # Is 'red' a valid color? True


invalid = Car.is_valid_color("purple")
# Is 'purple' a valid color? False
print(f"Is 'purple' a valid color? {invalid}")


# Can also call static methods from instance
print(toyota.calculate_car_age(2015))  # 11
````

**Why These Static Methods?**

| Static Method       | Why It's Static                                                                     |
| ------------------- | ----------------------------------------------------------------------------------- |
| calculate_car_age() | Doesn't need instance data or class data—just does math on a year                   |
| is_vintage_car()    | Pure utility function that classifies based on age, independent of any car instance |
| is_valid_color()    | Validates against a predefined list, not dependent on instance or class state       |

**Key Point:** Static methods are `utility functions` logically related to the class, but they don't need access to self (instance) or cls (class). You can call them on the class directly: `Car.calculate_car_age(2020)` or even on instances: `toyota.calculate_car_age(2020)`.

## OOP Principles

### Encapsulation

it involves bundling the data(attributes) and methods(functions) that operate on the data into a single unit, called a class.
Encapsulation is used to hide the internal details and implementation of the class from the outside world. It only exposes the necessary parts of the class to the outside world.

`we will only provide the getter method for the attribute, not the setter method.`

````python
class BankAccount:
    def __init__(self) -> None:
        self._balance: float = 0.0

    @property   # only provide getter method
    def balance(self) -> float:
        return self._balance

    def deposit(self, amount: float) -> None:
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self._balance += amount

    def withdraw(self, amount: float) -> None:
        if amount <= 0:
            raise ValueError("Withdraw amount must be positive.")
        if amount >= self._balance:
            raise ValueError("Insufficient funds.")
        self._balance -= amount
````

````python
chase_bank: BankAccount = BankAccount()
print(chase_bank.balance)
chase_bank.deposit(100)
print(chase_bank.balance)
chase_bank.withdraw(50)
print(chase_bank.balance)
````

### Abstraction

Reduce complexity by hiding unnecessary details.

Basically **encapsulation** is used to hide the internal details and implementation of the class from the outside world. and **abstraction** is used to reduce complexity by hiding unnecessary details.

````python
class EmailService:
    """Send emails using a simple connect-auth-send-disconnect workflow.

    This class demonstrates abstraction: callers only use `send_email()`, while
    connection and authentication details stay hidden in internal helper methods.

    Example:
        >>> service = EmailService()
        >>> service.send_email(
            to="john@gmail.com",
            subject="Welcome",
            body="Your account was created successfully.",
        )
    """

    def _connect(self) -> None:
        """Establish a connection to the email server.

        This internal helper method is part of the implementation detail and is
        not intended to be called directly by users of the class.
        """
        print("Connecting to email server")

    def _authenticate(self) -> None:
        """Authenticate with the email server.

        In a real implementation, this step would verify credentials or API tokens before sending messages.
        """
        print("Authenticating ... ")

    def _disconnect(self) -> None:
        """Close the email server connection.

        Keeping cleanup in a separate internal method makes the workflow easier to maintain and extend.
        """
        print("Disconnecting from email server")

    def send_email(self, to: str, subject: str, body: str) -> None:
        """Send an email message to one recipient.

        Args:
            to: Recipient email address.
            subject: Subject line of the email.
            body: Plain-text content of the email.

        Example:
            >>> service = EmailService()
            >>> service.send_email(
                to="john@gmail.com",
                subject="Hello",
                body="This is a test email.",
            )
        """
        self._connect()
        self._authenticate()
        print(
            f"Sending email to '{to}' with subject '{subject}' and body '{body}'")
        self._disconnect()
````

````python
gmail: EmailService = EmailService()
gmail.send_email("john@gmail.com", "Hello", "This is a test email.")
````

### Inheritance

**Inheritance** is a fundamental concept in object-oriented programming
(OOP) that involves **creating new classes** (subclasses or derived classes)
based on **existing classes** (superclasses or base classes).

````python
class Vehicle:
    def __init__(self, brand: str, model: str, color: str, year: int) -> None:
        self.brand: str = brand
        self.model: str = model
        self.color: str = color
        self.year: int = year

    def vehicle_start(self) -> None:
        print(f"The {self.brand} {self.model} is starting...")

    def vehicle_stop(self) -> None:
        print(f"The {self.brand} {self.model} is stopping...")
````

````python
class Car(Vehicle):
    def __init__(self, brand: str, model: str, color: str, year: int, num_seats: int) -> None:
        super().__init__(brand, model, color, year)
        self.num_seats: int = num_seats

    def __str__(self) -> str:
        return f"Car: {self.brand} {self.model} {self.color} {self.year} {self.num_seats}"
````

````python
raize: Car = Car("Toyota", "Corolla", "Red", 2020, 5)
print(raize)
raize.vehicle_start()  # inherit from Vehicle class
raize.vehicle_stop()  # inherit from Vehicle class
````

````python
class Motorcycle(Vehicle):
    def __init__(self, brand: str, model: str, color: str, year: int, engine_size: int) -> None:
        super().__init__(brand, model, color, year)
        self.engine_size: int = engine_size

    def __str__(self) -> str:
        return f"Motorcycle: {self.brand} {self.model} {self.color} {self.year} {self.engine_size}"
````

````python
bmw: Motorcycle = Motorcycle("BMW", "R1250GS", "Black", 2022, 1250)
print(bmw)
bmw.vehicle_start()  # inherit from Vehicle class
bmw.vehicle_stop()  # inherit from Vehicle class
````

````python
class Truck(Vehicle):
    def __init__(self, brand: str, model: str, color: str, year: int, num_seats: int) -> None:
        super().__init__(brand, model, color, year)
        self.num_seats: int = num_seats

    def __str__(self) -> str:
        return f"Truck: {self.brand} {self.model} {self.color} {self.year} {self.num_seats}"
````

````python
mac: Truck = Truck("Mercedes", "Actros", "White", 2021, 2)
print(mac)
````

### Polymorphism

The word polymorphism is derived from Greek, and means **"having multiple forms"**:

* Poly = many
* Morph = forms

In programming, polymorphism means that a class can have multiple forms.

````python
# Create list of vehicles to inspect

vehicles_list: list[Vehicle] = [
    Car("Ford", "Focus", "red", 2008, 5),
    Car("Toyota", "Corolla", "blue", 2010, 5),
    Motorcycle("Honda", "Scoopy", "black", 2018, 1250),
    Motorcycle("BMW", "R1250GS", "black", 2022, 1250),
    Truck("Mercedes", "Actros", "White", 2021, 2),
    Truck("Volvo", "FH", "Red", 2020, 2)
]
# Loop through list of vehicles and inspect them
for vehicle in vehicles_list:
    print(
        f"{vehicle.brand} {vehicle.model} is {type(vehicle).__name__}")
    vehicle.vehicle_start()
    vehicle.vehicle_stop()
    print("*"*25)
````
