#### notes

there is always a question. when we can generate random text and numbers with `random()` module then why we use `secrets()` module?

answer is simple. it is not cryptographically secure.

it is a terrible choice to use it for generating tokens and passwords. it gives a bad actor to predict what should be the next. because `random()` is not designed so.

so, if we are going to generate secure passwords, tokens or generate secure information, we should use `secrets()` module.

- You can generate truly secretive OTP, tokens, passwords and more!
- You can generate random data from a random module.
- It is not safe! it is not non-deterministic data.
- Data that is produced from the random modules can be determined easily by finding the seed that is used to produce the data.

```python
# let's create a random number from secretes module rather then random module.

import secrets

random_number: int = secrets.randbelow(100)

print(random_number)
```

```python
import secrets
import string

# https://docs.python.org/3/library/secrets.html#recipes-and-best-practices


def generate_password(length: int) -> None:
    """Print a cryptographically secure random password of the given length.

    Draws characters from the full ASCII alphabet (upper + lower), digits,
    and punctuation using ``secrets.choice``, which is backed by the OS
    CSPRNG and is safe for credential generation.

    Args:
        length: Number of characters in the generated password.
                Values below 8 are technically accepted but not recommended
                for real-world use.

    Example:
        >>> generate_password(16)
        Generated password: aB3$xQ!mZ8@kLp#2

    Note:
        The return type is ``None`` because the password is printed directly.
        Refactor to ``-> str`` and ``return password`` if you need to use the
        value programmatically.
    """
    chars: str = string.ascii_letters + string.digits + string.punctuation
    password: str = "".join(secrets.choice(chars) for _ in range(length))
    print(f"Generated password: {password}")


generate_password(15)
```

```python
# let's create a random number from given bitsize.

import secrets

random_number_upto_bits: int = secrets.randbits(16)

print(random_number_upto_bits)
```

`secrets.randbits(16)` asks the OS's **cryptographically secure random number generator** (CSPRNG — e.g. `CryptGenRandom` on Windows, `/dev/urandom` on Linux) for 16 random bits, then returns them as **an integer**.

What **"16 bits"** means in practice:

16 bits can represent **2¹⁶ = 65,536** possible values: **0 through 65535**
**Each bit** is **independently and unpredictably** 0 or 1, so the output is uniformly distributed across that range

Contrast with `secrets.randbelow(n)`:

|             | randbits(k)                 | randbelow(n)                |
| ----------- | --------------------------- | --------------------------- |
| Upper bound | Always a power of 2 (2ᵏ)    | Any integer n               |
| Use case    | Token generation, bit masks | Picking from an exact range |

**Why secrets instead of random.getrandbits(16)?**

random uses the **Mersenne Twister** — a fast **PRNG** whose internal state can be reconstructed after observing ~624 outputs. An attacker could predict future values.

secrets delegates to the OS `CSPRNG`, which is designed to be computationally infeasible to predict, making it appropriate for security-sensitive uses like session tokens or nonces.

```python
# let's create a random bytes tokens from given bytes

import secrets

random_tokens: bytes = secrets.token_bytes(32)

print(type(random_tokens))
print(random_tokens)
```

```python
# let's create a random hex token from given bytes
# since python version 3.11, it is 32  bytes
# https://docs.python.org/3/library/secrets.html#secrets.token_hex
# As of 2015, it is believed that 32 bytes (256 bits) of randomness is
# sufficient for the typical use-case expected for the secrets module

import secrets

random_tokens: str = secrets.token_hex(32)

print(type(random_tokens))
print(random_tokens)
```

```python
# let's create a random safe url token from given number
# it is best when you need your customer to provide with a secure random
# token, such as a password reset link or an email confirmation link.

# https://docs.python.org/3/library/secrets.html#secrets.token_urlsafe

import secrets

random_safe_url: str = secrets.token_urlsafe(8)

print(type(random_safe_url))
print(random_safe_url)
print(f"https://example.com/reset-password?token={random_safe_url}")

# python.org recommended code

# Generate a hard-to-guess temporary URL containing a security token suitable
# for password recovery applications:

url: str = "https://example.com/reset=" + secrets.token_urlsafe()
print(url)
```

```python
# sometime we need to compare username and password.
# we should not compare them the old way

userinput: str = "abc123456"
password: str = "abc123"

if userinput == password:
    print("Login successful!")
else:
    print("Login failed!")

# instead we should use secrets.compare_digest() to prevent against timing
# attacks
# https://docs.python.org/3/library/secrets.html#other-functions
# Return True if strings or bytes-like objects a and b are equal, otherwise
# False, using a “constant-time compare” to reduce the risk of timing attacks.

if secrets.compare_digest(userinput, password):
    print("Login successful!")
else:
    print("Login failed!")
```
