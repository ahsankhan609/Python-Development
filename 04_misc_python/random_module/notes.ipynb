#### Bookkeeping functions

```python
from random import randint, random, sample, seed

# https://docs.python.org/3/library/random.html#bookkeeping-functions
# Initialize the random number generator.
seed(42)


print(f"{random()=}")
print(f"{randint(1, 5)=}")
print(f"{sample(range(1000), k=5)=}")
```

#### Functions for integers

```python
import random

# https://docs.python.org/3/library/random.html#functions-for-integers

# Return a randomly selected element from range(start, stop, step).
# upper limit is exclusive
print(random.randrange(1, 10))  # 1 to 9

# Return a random integer N such that a <= N <= b. Alias for randrange(a, b+1).
# upper limit is inclusive
print(random.randint(1, 10))  # 1 to 10
```

#### 5 random useful use cases

```python
from random import shuffle

# let's say we have a list of people, and we need to shuffle them for a game.

people: list[str] = ["Alice", "Bob", "Charlie", "David", "Eve"]
print("Before shuffling:", people)
shuffle(people)
print("After shuffling:", people)
```

```python
from random import randint, random, randrange

# generate a random float between 0 to 1
value: float = random()
print(f"random() = {value}")

# upper limit is inclusive
values: list[int] = [randint(10, 20) for _ in range(5)]
print(f"randint() = {values}")

# upper limit is exclusive
values2: list[int] = [randrange(0, 3) for _ in range(5)]
print(f"randrange() = {values2}")
```

```python
from random import choice, choices

people: list[str] = ["Bob", "Tom", "James", "Sandra"]
print(f"choice() = {choice(people)}")

# be default its return 1 in list form
print(f"choices(return 1 in list format) = {choices(people)}")

# or we can specify N of choices
print(f"choices( return in list for N numbers) = {choices(people, k=2)}")

# we can also specify weight for each element to appear in the list
weights: list[float] = [0.15, 0.30, 0.25, 0.40]
print(
    f"choices( return in list for N numbers) = {choices(people, k=2, weights=weights)}")
```
