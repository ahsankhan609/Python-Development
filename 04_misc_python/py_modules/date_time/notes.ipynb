# Date and time Module

**Epoch :-** In computing, the epoch is often used as a starting point to calculate time intervals or timestamps.

`UNIX/windows :- January 1, 1970, at 00:00:00 UTC`

this above value is system based. it can be negative or positive.

- positive values represent time after the epoch.
- negative values represent time before the epoch.

### time module

The time module in Python provides various functions to work with time-related operations.

Below are important functions from time module :-
- time() - returns(int value) the **current time in seconds since Epoch**. January 01, 1970 at 00:00:00
- sleep() - **suspends** the **execution of current thread** for specified number of **seconds**
- ctime() - converts a timestamp(seconds since the epoch) into **human-readable string**, representing local time.
- gmtime() - converts a timestamp into a **struct_time object** representing UTC time
- localtime() - converts a timestamp into a struct_time object representing local timezone of the system
- strftime() - converts a struct_time object into a **formatted string** based on specified format codes
- strptime()
- monotonic()

````python
import time

print(f"time from Epoch is: {time.time()} in seconds.")
````

````python
time_start: int = time.time()
print("hello world")
time_end: int = time.time()

print(f"seconds taken to print: {time_end - time_start}")
````

````python
time.sleep(10)
print("printed after sleep.")
````

````python
# return string representation of current local time Since Epoch
print(time.ctime())
# return string representation of time 1000 seconds ahead of Epoch
print(time.ctime(1000))
````

````python
# return struct_time object representation of current local time Since Epoch
print(time.gmtime())

# return struct_time object representation of current local time after
# 1000 seconds of Epoch
print(time.localtime(1000))

# we can access the struct_time object attributes like this
current_time_struct: time.struct_time = time.localtime()
print(f"current year is: {current_time_struct.tm_year}")
print(f"current month is: {current_time_struct.tm_mon}")
````

````python
print(current_time_struct)
````

20. -----------------------------------------------------------------

### datetime module with classes and methods

There are some limitations in the time module. we can only work with the time in a limited way. The datetime module in Python provides classes for working with dates, times, and combinations of dates and times.

Datetime module has multiple classes:-

1) datetime :- include both date and time components

2) date :- When you need to work with dates without
considering the time component

3) time :- For representing specific times of the day
without considering dates

4) timedelta :- used to represent differences or durations
between two datetime objects.

import a module named **datetime** to work with dates and time as **datetime object**.

```python
import datetime as dt

x = dt.datetime.now(tz=dt.UTC)

print(x)
```

The date object contains year, month, day, hour, minute, second, and microsecond.

````python
import datetime as dt

today: dt.datetime = dt.datetime.now(tz=dt.UTC)
# returns current date and time object in UTC timezone

only_time: dt.time = dt.datetime.now(tz=dt.UTC).time()
# returns current time object in UTC timezone

month: int = today.month  # returns month of the year as an integer value
year: int = today.year  # returns year as an integer value
timezone: dt.timezone = dt.UTC
# returns timezone information of the datetime object.
# In this case, it will return UTC timezone information.

print(today)
print(only_time)
print(today.day)
print(month)
print(year)
print(timezone)
print(today.strftime("%B %d, %Y %H:%M:%S"))
````

#### Use Case 1: Free Trial Expiry Checker

````python
# https://claude.ai/share/98829c8c-79b9-4799-b60f-7254f906d165
from datetime import datetime, timedelta


def check_trial_status(signup_date: datetime, trial_days: int = 14) -> dict:
    """
    Check if a user's free trial is still active or expired.

    Args:
        signup_date: The date when the user signed up.
        trial_days: Number of free trial days (default: 14).

    Returns:
        A dict containing trial status, expiry date, and remaining days.
    """
    trial_expiry: datetime = signup_date + timedelta(days=trial_days)
    today: datetime = datetime.now()
    remaining: timedelta = trial_expiry - today

    return {
        "signup_date": signup_date.strftime("%Y-%m-%d"),
        "expiry_date": trial_expiry.strftime("%Y-%m-%d"),
        "is_active": today < trial_expiry,
        "days_remaining": max(remaining.days, 0),
    }


# --- Driver Code ---
signup = datetime(2026, 5, 15)
status = check_trial_status(signup)

print(f"Trial Active  : {status['is_active']}")
print(f"Expiry Date   : {status['expiry_date']}")
print(f"Days Remaining: {status['days_remaining']}")
````
