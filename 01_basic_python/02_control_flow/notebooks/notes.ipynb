**Replacing Messy if-elif Chains.**

Here we will demonstrate how to replace messy and bulky if-elif chains with a more
cleaner, readable and maintainable code. we will use a dictionary to map
user roles to their access levels.

```python
def process(user_role: str) -> str:
    """this function will return the access level for a given user role."""

    if user_role == "admin":
        return "Access granted: Full"
    elif user_role == "editor":
        return "Access granted: Partial"
    elif user_role == "viewer":
        return "Access granted: Read-only"
    else:
        return "Access denied"
```

```python
def process(user_role):
    return access_map.get(user_role, default_access)()
```

**Why This Pattern Works So Well**

**1. Scales Without Pain**
Adding a new role doesn’t require touching existing logic. You simply define a new function and update the dictionary.

**2. Cleaner Separation of Concerns**
Each behavior lives in its own function. This makes your code easier to read, easier to reason about, and much easier to unit test.

**3. Better Performance Than Long Conditionals**
Python dictionaries are highly optimized. A direct lookup is often faster than evaluating multiple conditional expressions.

**4. Supports Dynamic and Config-Driven Logic**
You can populate the function map at runtime — from configuration files, plugins, or even user input.

**Real-World Use Cases**
This pattern shines in many practical scenarios:

* Command handlers in CLI tools or chatbots
* Role-based access control
* State machines and workflow engines
* Form validation, mapping field types to validators
* Factory patterns, dynamically creating objects or behaviors
* Anywhere you see repeated `if condition == value` logic, this approach is worth considering.

**What If Your Functions Need Arguments?**

No problem — this pattern works seamlessly with parameters.

```python
def greet_en(name: str) -> str:
    return f"Hello, {name}!"


def greet_es(name: str) -> str:
    return f"¡Hola, {name}!"


def greet_fr(name: str) -> str:
    return f"Bonjour, {name}!"


greet_map: dict[str, Callable[[str], str]] = {
    "en": greet_en,
    "es": greet_es,
    "fr": greet_fr,
}


def greet(language: str, name: str) -> str:
    return greet_map.get(language, greet_en)(name)
```
