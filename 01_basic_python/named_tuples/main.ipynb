Python’s `namedtuple` in the `collections module` allows you to `create immutable sequences with named fields`, providing a `more readable` and `Pythonic` way to handle tuples. You use `namedtuple` to `access values with descriptive field names` and dot notation, which improves code clarity and maintainability.

* it is a factory function, from collections module, which becomes sub-class of tuple.
* The `main difference` between `tuple and namedtuple` is that namedtuple `allows attribute access via named fields`, enhancing readability.
* In original tuple we have no field names, but here we have.
* This is the benefit of using namedtuple, to improve code clarity by `allowing access to elements through descriptive names instead of integer indices`.
* namedtuple is `immutable`, just like regular tuples, which means you cannot modify the values once they are set. This immutability makes named tuples a great choice for representing fixed data structures.

Problem:

In contrast, accessing values by index in a regular tuple can be frustrating, hard to read, and error-prone. This is especially true if the tuple has a lot of fields and is constructed far away from where you’re using it.

summary of how each term is used throughout the tutorial:
* namedtuple() - The factory function for creating named tuples.
* namedtuple, namedtuple -> class - The tuple subclass returned by namedtuple()
* namedtuple instance, named tuple - An instance of a specific namedtuple class

Besides providing named fields, named tuples in Python offer the following features:

* Are immutable data structures
* Can have a `hash value` and work as `dictionary keys`
* Can be stored in `sets` when they have a `hash value`
* Generate a basic `docstring` using the type and field names
* Provide a helpful `string representation` that displays the tuple content in a name=value format
* Support `indexing` and `slicing`
* Provide additional methods and attributes, such as ._make(), _asdict(), and ._fields
* Are `backward compatible with regular tuples`
* Have `similar memory usage` to regular tuples

### Creating Tuple-Like Classes With the `namedtuple()` Function

````python
# Create a 2D point as a regular tuple
point = (2, 4)
print(point)  # (2, 4)

# Access coordinate x
print(point[0])  # 2

# Access coordinate y
print(point[1])  # 4

# Try to update a coordinate value
try:
    point[0] = 100  # it will create error
except:
    print("The point is immutable, so you can’t modify the coordinates")
````

However, do you think this code is readable? Can you tell upfront what the `0` and `1` indices mean?

To `improve clarity`, you can use a `namedtuple`. Note that you need to `import the function` from the `collections` module first:

````python
from collections import namedtuple

# Create a namedtuple type, Point
Point = namedtuple("Point", "x y")

point = Point(2, 4)
print(point)  # Point(x=2, y=4)

# Access the coordinates by field name
print(point.x)
print(point.y)

# we can also Access the coordinates by index, if we prefer
point[0]
point[1]

# Try to update value of Point x, we can't update it
try:
    point.x = 100
except:
    print("The point is immutable, so you can’t modify the coordinates")

print("Is Point is subclass of tuple(by using issubclass): ",
      issubclass(Point, tuple))
````

Now we have a `Point class` with two `appropriately named fields`, `x` and `y`. Ou Point Class provides a descriptive string representation by default: `Point(x=2, y=4)`.

`Note:`

As with regular tuples, named tuples are immutable. However, the values they store don’t necessarily have to be immutable.

It’s completely valid to create a tuple or a named tuple that holds mutable values:

````python
from collections import namedtuple

# let's create a namedtuple with immutable capability

# Person = namedtuple("Person", "name Children wives")# can be write by this

# can be write by this
Person = namedtuple("Person", ["name", "children", "wives", "parent"])
# when we create instance of the Person, the typename field will be filled with that name of the instance that we create, for example John in example below


# let's add mutuable values in the named tuple
john = Person("John Doe", ["Timmy", "Jimmy"],
              ["sara", "sasha"],
              ["Martha", "Richard"])
print(john)

print("ID of Childre: ", id(john.children))
print("ID of Wives: ", id(john.wives))
print("ID of Parents: ", id(john.parent))

# now let's try to append another children to the family
print("Added new child to the family.")
john.children.append("Tina")
print(john)
print("ID of Childre: ", id(john.children))

try:
    hash(john)
except:
    print("Finally, tuples or named tuples with mutable values aren’t hashable, as you saw in the above example.")
````

we can create `named tuples` that `contain mutable objects`. Then, we can `modify the mutable objects` in the underlying tuple. However, this `doesn’t mean that you’re modifying the tuple itself`. The tuple will continue `being the same object`.

#### Explanation of the Above code (brief)
- `Person = namedtuple("Person", ["name", "children", "wives", "parent"])` creates an immutable tuple subclass named Person with four fields.
- `john = Person(...)` constructs an instance where fields hold the provided values (strings and lists).
- `print(john)` shows the Person representation.
- `id(...)` calls print the object identity (memory address) of each field value (lists here).
- `hash(john)` is attempted inside try/except; it raises an exception because the instance is unhashable in this case, so the except branch prints the explanatory message.

1 — What "hashable" means for namedtuples
- Hashable means an object has a stable integer hash value returned by hash(obj) and can be used as a key in dicts or as an element of a set.
- For compound objects (like tuples or namedtuples) Python considers them hashable only if all their constituent parts are themselves hashable and immutable. The object's __hash__ must be defined and produce a value that doesn't change for the object's lifetime.

2 — Why this Person instance is not hashable
- The namedtuple class itself provides tuple-like behavior and normally is hashable if its fields are hashable.
- In this example, some fields (`children`, `wives`, `parent`) are Python lists, which are mutable and therefore unhashable.
- Because the namedtuple contains unhashable components, computing hash(john) fails (Python raises TypeError), making the instance effectively unhashable.
- The printed ids show the lists are distinct mutable objects (their identities are stable but their contents can change), which is why they break hashability.

If you need a hashable Person, use only immutable, hashable field values (e.g., use tuples instead of lists for those fields).

### Let's See example of Hasable Namedtuple:

````python
# Use only immutable, hashable field values (e.g., strings and tuples) so the namedtuple instance can be hashed.

from collections import namedtuple

Person = namedtuple("Person", ["name", "children", "wives", "parent"])

john = Person(
    "John Doe",
    ("Timmy", "Jimmy"),      # tuple (immutable, hashable)
    ("sara", "sasha"),       # tuple
    ("Martha", "Richard")    # tuple
)

print(john)
print("Hash of the Person namedtuple subclass, becuae we use immutable fields:", hash(john))
print("Can be dict key:", {john: "some value"})
````

### Supplying Required Arguments

To create a new namedtuple class, we need to provide two positional arguments to the function:

1. `typename` provides the class name for the namedtuple class returned by `namedtuple()`. You need to pass a string with a valid Python identifier to this argument.
2. `field_names` provides the field names that you’ll use to `access the values in the tuple`. You can `provide the field names` using `one` of the following options:
   * An iterable of strings, such as ["field1", "field2", ..., "fieldN"]
   * A string with each field name separated by whitespace, such as "field1 field2 ... fieldN"
   * A string with each field name separated by commas, such as "field1, field2, ..., fieldN"

```python
Person = namedtuple("Person", ["name", "children", "wives", "parent"])
Person = namedtuple("Person", "name children wives parent")
Person = namedtuple("Person", "name, children, wives, parent")
```

The `typename` argument is `required` because it `defines the name of the new class` being created. It’s similar to `assigning a name to a class` definition manually, as shown in the following comparison:

````python
from collections import namedtuple

Point1 = namedtuple("Point", "x y")
print(Point1)


class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y


point2 = Point(2, 4)
print(point2)
````

### how different named tuples and dictionaries are in terms of performance.

````python
from collections import namedtuple
from time import perf_counter


def average_time(structure, test_func):
    time_measurements = []
    for _ in range(1_000_000):
        start = perf_counter()
        test_func(structure)
        end = perf_counter()
        time_measurements.append(end - start)
    return sum(time_measurements) / len(time_measurements) * int(1e9)


def time_dict(dictionary):
    "x" in dictionary
    "missing_key" in dictionary
    2 in dictionary.values()
    "missing_value" in dictionary.values()
    dictionary["y"]


def time_namedtuple(named_tuple):
    "x" in named_tuple._fields
    "missing_field" in named_tuple._fields
    2 in named_tuple
    "missing_value" in named_tuple
    named_tuple.y
````

````python
Point = namedtuple("Point", "x y z")
point = Point(x=1, y=2, z=3)

namedtuple_time = average_time(point, time_namedtuple)
dict_time = average_time(point._asdict(), time_dict)
gain = dict_time / namedtuple_time

print(f"namedtuple: {namedtuple_time:.2f} ns ({gain:.2f}x faster)")
print(f"dict:       {dict_time:.2f} ns")
````

This output shows that operations on named tuples are slightly faster than similar operations on dictionaries.

### namedtuple vs Data Classes

Python comes with data classes, which—according to PEP 557—are similar to named tuples but are mutable.

However, it’d be more accurate to say that data classes are like mutable named tuples with type hints. The defaults part isn’t a real difference because named tuples can also have default values for their fields. So, at first glance, the main differences are mutability and type hints.

````python
from dataclasses import dataclass


@dataclass
class Person:
    name: str
    age: int
    height: float
    weight: float
    country: str = "Canada"


jane = Person("Jane", 25, 1.75, 67)
print(jane)


Person(name='Jane', age=25, height=1.75, weight=67, country='Canada')
print(jane.name)


jane.name = "Jane Doe"
print(jane.name)
````

In terms of readability, there are no significant differences between data classes and named tuples. Both provide similar string representations and allow attribute access using dot notation.

When it comes to mutability, data classes are mutable by definition, so you can change the value of their attributes when needed. However, they have an ace up their sleeve. You can set the @dataclass decorator’s frozen argument to True to make them immutable

````python
from dataclasses import dataclass


@dataclass(frozen=True)
class Person:
    name: str
    age: int
    height: float
    weight: float
    country: str = "Canada"


jane = Person("Jane", 25, 1.75, 67)
print(jane)

try:
    Person(name='Jane', age=25, height=1.75, weight=67, country='Canada')
    jane.name = "Jane Doe"
    print(jane.name)
except Exception as e:
    print("Your have : ", type(e).__name__, ":", e)
````
