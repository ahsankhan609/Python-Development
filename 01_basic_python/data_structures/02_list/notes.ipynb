# Python Lists


- Lists are used to `store multiple data type items` in `a single variable`
- `Lists and Arrays` are `Same`. But there `are some differences`.
- Lists are created using `square brackets` like `[]`
- It allows Duplicate items with the same value.
- Arrays are homogeneous. Means, it `store same Data type` in it. Data `stores on the same memory location`. Arrays are `much faster` for computation because of data at same memory location.
- Lists are heterogeneous. Means we `can store multiple Data types`.
- In Lists `data` `stored on different memory locations`.
- Lists are `slower for computation`.
- Lists are more programmer friendly.
- [Python List Methods - w3School](https://www.w3schools.com/python/python_lists_methods.asp)


```python
# Example of List with Mutiple data types
thislist:list = ["apple", "banana", "cherry", 1, 2, 3, True, 4+5j]
print(thislist)
```

```python
# Another example of 2d Hetrogenous List
this_list:list = [1, 2, 3, [4, 5, 6]]
print(this_list)
```

```python
# The list() Constructor
thislist:list = list(("apple", "banana", "cherry")) # note the double round-brackets
print(thislist)
```

#### Access List Items

```python
l3:list = [1, 2, 3, [4, 5, 6]]

# Example 1
print(l3[3][0])

#Example 2
print(l3[-1][0])

# Negative indexing `start from the end`

# `-1` refers to the last item, `-2` refers to the second last item etc.

thislist:list[str] = ["apple", "banana", "cherry"]
print(thislist[-1])
```

#### Range of Negative Indexes

```python
# Specify negative indexes if you want to start the search from the end of the 02_list:

# This example returns the items from "orange" (-4) to, but NOT including "mango" (-1):

thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[-4:-1])
```

#### Append into List

```python
l:list[int] = [1, 2, 3, 4]
l.append(5)
print(l)
```

#### Extend into List - Combine 2 lists

```python
l:list[int] = [1, 2, 3, 4]
l.extend([6, 7, 8])
print(l)
```

Difference Between Append and Extend is:-

 - Both add items at the end of List.
  - Append just add 1 Item,
     - Extend add multiple Items.

```python
# So for example if we try to append multiple items, then it will add that item as 1
l.append([9, 10])
print(l) # so it add 02_list as 1 item

# So for example if we try to extend single item, then it will add it  as multiple items
l.extend('islam')
print(l)
```

#### Insert Items in List

```python
# To insert a new item, without replacing any of the existing values, we can use the insert() method.
# The insert() method inserts an item at the specified index:

thislist:list[str] = ["apple", "banana", "cherry"]
thislist.insert(1, "watermelon")
print(thislist)
```

#### Remove Specified Item from List

```python
# The remove() method removes the specified item.
thislist:list[str] = ["apple", "banana", "cherry"]
thislist.remove("banana")
print(thislist)

# If there are more than one item with the specified value, the remove() method removes the first occurrence:
thislist = ["apple", "banana", "cherry", "banana", "kiwi"]
thislist.remove("banana")
print(thislist)

# Remove Specified Item from List - The pop() method removes the specified index.
thislist = ["apple", "banana", "cherry"]
thislist.pop(1)
print(thislist)

# If you do not specify the index, the pop() method removes the last item.

thislist = ["apple", "banana", "cherry"]
thislist.pop()
print(thislist)
```
