# Sets

- empty Set is only made like this - set_name = set()
- initiate with {} - s = {101,'john',20.20, True}
- OR - s = set(iterable)
- it is used to store un-ordered collectin of data.
- it can be integer, boolean, string, float
- don't allow duplicate values.
- indexing and slicing cannot be implemented
- Sets are Mutable - we can add value to it, it will be added at any index
- elements in Set must be immutable

```python
my_set_1 = {101, 'john',20.20, True, 'john', 20.20}
my_set_2 = set([101,'john', 20.25, False, 101,101, False])

print(my_set_1)
print(my_set_2)
```

#### Builtin Methods

```python
# add() - add one element to the set at any random index position
my_set_1.add(102)
print(my_set_1)
```

```python
# update() - add multiple element to the set at any random index positions
my_set_1.update([103,104, False, True,75])
print(my_set_1)
```

```python
# remove() - remove one element from the set, if again try to remove old element, it will throw error
my_set_1.remove(103)
print(my_set_1)
```

```python
# discard() - remove one element from the set, but if try to remove old element, it will not throw error
my_set_1.discard(103)
print(my_set_1)
```

```python
# pop() - remove any random element from the set
my_set_1.pop()
print(my_set_1)
```

```python
# clear() - remove all elements from the set and returns empty set if checked
my_set_1.clear()
my_set_1
```

```python
# del - it will delete the set from memory, if called it will throw error
del my_set_1
print(my_set_1)
```

#### Builtin Operations

Union of two given sets is the smallest set which **contains all the unique elements** from both of the sets.

```python
# Union of Sets
A = {2, 4, 5, 6,7,8,9}
B = {1,3,4, 6, 7, 8,9}
print(A.union(B))
print(A | B)
```

Intersection of two given sets is the largest set which contains all the **elements that are common** to both the sets.

```python
# Intersection of Sets
A = {2, 4, 5, 6}
B = {4, 6, 7, 8}
print(A.intersection(B))
print(A & B)
```

The difference between the two sets is equal to the difference between the number of elements in two sets.

```python
# Difference of Sets
A = {2, 4, 5, 6}
B = {4, 6, 7, 8}
print(A.difference(B))
print(A - B)
print(B - A)
```

The symmetric difference of two sets is a set of **elements that are in either set**, but not in their intersection.

```python
# Symmetric difference of Sets
A = {2, 4, 5, 6}
B = {4, 6, 7, 8}
print(A.symmetric_difference(B))
print(A ^ B)
```
