# Tuple

#### Examples

```python
# tuples are hashable

# let's say we need to keep unique coordinates without any duplicates

cords = set()

x = (1,2)
y = (1,2)
cords.add(x)
cords.add(y)
print(cords)
```

```python
# memory sharing
print(id(x))
print(id(y))
```

```python
# unpacking
a,b,c = (1,2,3)
print(a,b,c)

# packing
nums = a,b,c
print(nums)
```
