```python
var1 = 10
var2 = 'john smith'

print(f'{var2} is {var1} years old. and there are {30*24} hours in month.')
print(f"{var2.upper()}")
```

```python
# by now we know that if we have long number, we use it like this 1_000_000_000 and interpreter ignore it.
# it's only for the developer to understand it

number:int = 1_000_000_000_000
number_with_scientific_notation:float = 1e9 # nine zeros after 1
print(number)
print(f"{number:_}")
print(f"{number:,}")
print(number_with_scientific_notation)
```

```python
big_number:int = 2_650_000_000_000
print(f"{big_number:.3e}") # it means 3 float numbers after . and then e represents number of Spaces after 1st number
```

```python
var:str = "var"
print(f"{var:_>20}:") # it will print 17 spaces to left side of variable, and we can fill space with symbols
print(f"{var:#<20}:") # by default we have right space, so we can remove it too, and we can fill space with symbols
print(f"{var:|^20}:") # it will center align it, and we can fill space with symbols
```

```python
from datetime import datetime as dt

today_date:dt = dt.now()
print(f"{today_date: %B %d, %Y}")
print(f"{today_date: %c}") # return local version
```

```python
# nesting variables with f-strings

date_spec:str = "%m %d, %Y"
print(f"{now:{date_spec}}")
```

```python
decimal_number:float = 120563.632514
print(f"{decimal_number:.2f}")
print(f"{decimal_number:,.2f}")
print(f"{decimal_number:_.2f}")
```

```python
another_decimal_number: float = 1000000.1234567
spec: str = ',.2f'
print(f' {another_decimal_number : {spec}}')
```

```python
a: int = 5
b: int = 10

my_var: str = 'Bob says hi'
my_path:str = r'\Users\federico\Documentsfolder'

print(f'a + b = {a + b}') # normal approach
print(f"{a + b = }") # moder approach
print(f"{my_var = }") # modern approach

print(f"{a + b = :.1f}") # modern approach with different formatters
print(f"{my_var = !s}") # modern approach, with telling interpreter that consider it as string. it will print without ''
print(f"{my_var = !r}") # modern approach, with telling interpreter that consider it as representation of string(__repr__). it will print with ''
```

```python
from datetime import datetime, date
banana: str = '🍌'
name: str = 'Bob'
today: date = datetime.now().date()

print(f'[{today!s}] {name!s} says: {banana!s}') # string representation of each variable
print(f'[{today!r}] {name!r} says: {banana!r}') # representation of each variable
print(f'[{today!a}] {name!a} says: {banana!a}') # ascii representation of each variable
```

```python
number_representation:int = 195

print(f"{number_representation:b} | Binary representation")
print(f"{number_representation:o} | Octal representation")
print(f"{number_representation:x} | Hexadecimal (lowercase) representation")
print(f"{number_representation:X}| Hexadecimal (uppercase) representation")
print(f" {number_representation:#x} | Hexadecimal representation with prefix")

percentage:float = 0.876
print(f"{percentage :.2%} | Percentage format with 2 decimal places")

other_number = 12345.6789
print(f"{other_number :.2f} | Two decimal places")
print(f"{other_number:,.2f} | Comma as thousands separator, 2 decimal places")
print(f"{other_number:010.2f} | Zero-padded width 10")
print(f"{other_number:+.2f} | Show sign(+) before number")
print(f"{other_number:011} | assure total number of chars should be equal to specified number.")
```

```python
# helpful in debugging
debug_n = 256.6328
print(f"DEBUGGING {debug_n}")
print(f"DEBUGGING {debug_n = }")
print(f"DEBUGGING {debug_n ** 2 = }")
print(f"DEBUGGING {debug_n ** 2 =: ,}")
print(f"DEBUGGING {debug_n ** 2 =: ,.3f}")
```

```python
from dataclasses import dataclass

@dataclass
class User:
    name: str
    age: int
    weight:float

john:User = User('John',25, 96.52)
print(john) # by default, it generates the __repr__ class for developer friendliness
```
