# frozen sets

- frozen set are immutable
- while elements of set can be modified, but frozenset elements remain same after creation
- don't allow duplicate values
- can't add, update or remove objects from it
- but we can do operations such as union, intersection etc...
- works on iterables

```python
# works on iterables
my_list:list[str] = ["python", "python", "djanog", "flask", "react", "js", "js"]

# create an instance of frozen set
fset:frozenset = frozenset(my_list)
print(fset)

# check type
print(type(fset))

# we can't add anything
try:
    fset.add("flask")
    print(fset)
except Exception as e:
    print(e)
```

```python
fset1:frozenset = frozenset([4, 5, 6, 7,8,9,10,10,12,85,54,14,10])
fset2 = frozenset([1, 2, 3, 4, 5, 6,10,14])
fset3 = fset1.union(fset2)
print(fset3) # return unique values from both sets
```

```python
# application of frozen set
week:frozenset[str] = frozenset(['mon','tue','wed','thu','fri','sat','sun'])
vowels:frozenset[str] = frozenset(['a','e','i','o','u'])
```

```python
# Using frozenset as Dictionary Keys

# Tracking user permissions by role
permissions:dict = {
    frozenset(['read', 'write']): 'editor',
    frozenset(['read']): 'viewer',
    frozenset(['read', 'write', 'delete', 'admin']): 'administrator'
}

user_perms:frozenset[str] = frozenset(['read', 'write'])
print(permissions[user_perms])  # Output: editor
```
