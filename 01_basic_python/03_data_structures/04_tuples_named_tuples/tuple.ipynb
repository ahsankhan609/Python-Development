# Tuple

- it will be ordered - no matter how long you run program, it will follow the order
- it will be like this (,)
- it is immutable - we can't add or remove any element from it
- it allowed duplicates
- it allowed any data types

```python
people:tuple = ('ali','hamza','ramzan','salman', 'hamza')
single_tuple:tuple = ('ahmed',)
mixed_tuple:tuple = ('a','b','c', 10 ,20 ,30, 4+6j, True, False)

print(type(people))
print(type(single_tuple))
print(type(mixed_tuple))
```

#### accessing elements with indexing

```python
print(people[2])
```

#### conditions

```python
print('hamza' in people)
```

#### converting to list

```python
people_list:list = list(people) # now we can perform list operations on it
print(type(people_list))
print(people_list)
```

#### count how many times an element exist

```python
print(people.count('hamza'))
```

#### find index of element - returns first occurrence

```python
print(people.index('hamza'))
```

#### Examples

```python
# tuples are hashable

# let's say we need to keep unique coordinates without any duplicates

cords = set()

x = (1,2)
y = (1,2)
cords.add(x)
cords.add(y)
print(cords)
```

```python
# memory sharing
print(id(x))
print(id(y))
```

```python
# unpacking
a,b,c = (1,2,3)
print(a,b,c)

# another way of unpacking
a, *b = (1,2,3)
print(b) # it will return list of remaining elements

# packing
nums = a,b,c
print(nums)
```
