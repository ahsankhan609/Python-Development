#### Catch MUltiple Errors and `finaly block`

we can catch multiple errors individually. But it is `not a good practice`. `best way` is to `catch errors as exceptions` and then check the type of error. And by default only first error is displayed.

So, in the case where we `need to catch multiple errors`, then we should `always use Exception as e in the end`.

```python
try:
    numenator: int = int(input("Write numenator: "))
    denominator: int = int(input("Write denominator: "))
    result: float = numenator / denominator
    print(result)

    my_list: list[int] = [1, 2, 3]
    i = int(input("Enter index: "))
    print(my_list[i])

except ZeroDivisionError:
    print("Denominator cannot be 0. Please try again.")
except ValueError:
    print("Denominator should be integer. Please try again.")
except IndexError:
    print("Index cannot be greater than size of list.")
finally:  # it will always run whatever the output is. Ususally it is used to close the open file, whether if the program occurs error
    print("Thanks for enjoying program")
```

if we catch Exception, it means that we can catch all the children of exception class. But if we have only catched `ZeroDivisionError` then we only get that result.

```python
# Parent class: Polygon
class Polygon:
    def __init__(self, sides: int) -> None:
        self.sides: int = sides

    def show_sides(self) -> None:
        print(f"This polygon has {self.sides} sides.")

# Child class: Triangle


class Triangle(Polygon):
    def __init__(self) -> None:
        super().__init__(3)  # A triangle has 3 sides

# Child class: Square


class Square(Polygon):
    def __init__(self) -> None:
        super().__init__(4)  # A square has 4 sides
```

```python
# Function to demonstrate casting
def check_polygon_type(polygon):
    if isinstance(polygon, Polygon):  # Checking if it's a parent class (Potygon)
        # print("True")
        polygon.show_sides()
    else:
        print("Not a valid polygon.")
```

```python
tri = Triangle()
sq = Square()

# Casting child to parent (allowed)
check_polygon_type(tri)  # Triangle is cast to Polygon
check_polygon_type(sq)  # Square is cast to Polygon
```

```python
# Function to demonstrate casting
def check_square_type(square) -> None:
    if isinstance(square, Square):  # Checking if it's a parent class (Polygon)
        square.show_sides()
    else:
        print("Not a valid Square.")
```

```python
# Creating instances of child classes

pol = Polygon(5)

check_square_type(pol)

# Attempt to cast a general Polygon to Triangle (not allowed directly)
poly = Polygon(5)
# check_polygon_type(Triangle(poly)) # This would raise an error (uncomment to see
```

### Else Block

* it executed when `try` block doesn't raise any exception.
* it is a good place to put code that should be executed when there is no error.

```python
def main() -> None:
    a = 10.0
    try:
        user_input: str = input(
            "Enter a number to use for division and addition with 10: ").strip()
        b = float(user_input)
        total: float = a + b
        quotient: float = a / b

    except ValueError:
        print("Invalid input: please enter a numeric value (e.g. 3.14 or 5).")
    except ZeroDivisionError:
        print("Math error: division by zero is not allowed—enter a number other than 0.")
    except Exception as e:
        print(f"Unexpected error: {e!s}")
    else:
        print(f"Success! 10 ÷ {b} = {quotient}")
        print(f"10 + {b} = {total}")
    finally:
        print("Program finished.")


if __name__ == "__main__":
    main()
```

#### Better Code explaination

* use `try block` to get user input, and perform operation on it.
* then handle `all possible exceptions` with proper message, and Final exception at then end
* then use `else block` towrite your code results that are ok, and should be return to the user, if there is no error
* then use `finally block` to write thank you message. for example: above you perform som operations on file, then its good place to close that file.

### finally Block

* it is always executed, whether an exception is raised or not.
* it is often used to perform cleanup actions, such as closing files or releasing resources, regardless of whether an error occured or not.

```python
class DB:
    def conn(self) -> None:
        print("connected")

    def close(self) -> None:
        print("connection closed")

    def get(self) -> None:
        print("here are your 500 records")
```

```python
try:
    db = DB()
    db.conn()
    db.get()
except ValueError as e:
    print(f"Error Occured as {e}")

finally:
    db.close()
```

```python
def ReadingFiles():
    try:
        print("I am here, opening all the files")
        return "Success, got all files"
    except ZeroDivisionError as e:
        print(e)
    finally:
        print("Closing all the files")


# here we can see that finally is so much powerful, that it will run first, whether we return something or not.
ReadingFiles()
```

### Raise - raise our own exceptions

* python allows us to manually raise exceptions using the `raise` keyword.
* useful when we want to enforce certain conditions in our code. For example, if we want to ensure that a function only accepts positive numbers, we can raise a `ValueError` if the input is negative.

```python
class NegativeValueError(ValueError):
    """Raised when a negative value is provided where only non-negative is allowed."""
    pass


def main() -> None:
    a = 10.0
    try:
        user_input: str = input(
            "Enter a non-negative number to use for division and addition with 10: ").strip()
        b = float(user_input)

        if b < 0:
            raise NegativeValueError("Negative values are not allowed.")

        total: float = a + b
        quotient: float = a / b

    except NegativeValueError:
        print("Input error: please provide a non-negative number (0 or greater).")
    except ValueError:
        print("Invalid input: please enter a numeric value (e.g. 3.14 or 5).")
    except ZeroDivisionError:
        print("Math error: division by zero is not allowed—enter a number other than 0.")
    except Exception as e:
        print(f"Unexpected error: {e!s}")
    else:
        print(f"Success! 10 ÷ {b} = {quotient}")
        print(f"10 + {b} = {total}")
    finally:
        print("Program finished.")


if __name__ == "__main__":
    main()
```

### Custom Errors

* Python allows us to create our Own `Custom Exception` by creating a new class

* allows us to create meaningful error messages specific to opplication / use case
* To handle custom exceptions:- 
  * Make our own class
  * inherit from Exception class

```python
class WrongEmail(Exception):
    pass
```

```python
class Person:
    def __init__(self, salary: int, email: str) -> None:
        if (salary <= 0):
            raise ValueError("Please enter a valid Salary")
        self. salary: int = salary

        if ('@' not in email):
            raise WrongEmail("please enter a valid Email")
        self.email: str = email

    def show_salary(self) -> int:
        # print(self.salary)
        return self.salary

    def show_email(self) -> str:
        # print(self.email)
        return self.email
```

```python
try:
    p1 = Person(100000, "abc@test.com")
    p2 = Person(100000, "xyz@test.com")

except ValueError as e:
    print(f"Please follow it: {e}")
except WrongEmail as e:
    print(f"Please follow it: {e}")

else:
    print(f"Your Salry is {p1.show_salary()}, and email is {p1.show_email()}")
    print(f"Your Salry is {p2.show_salary()}, and email is {p2.show_email()}")

finally:
    print("thank you for using our program.")
```

#### Define custom exception

```python
# Define custom exceptions
class InvalidAgeError(Exception):
    """Exception raised for invalid age."""

    def __init__(self, age: int):
        self.age: int = age
        super() .__init__(
            f"Invalid age: {age}. Age must be a positive integer.")


class InvalidEmailError(Exception):
    """Exception raised for invalid email format."""

    def __init__(self, email: str):
        self.email: str = email
        super() .__init__(
            f"Invalid email format: {email}. Email must be in the format 'username@example'")
```

```python
# define the person class

class Person:
    def __init__(self, name: str, age: int, email: str):
        self.name: int = name

        # Validate age
        if age <= 0:
            raise InvalidAgeError(age)
        self.age: int = age

        # Validate email format
        if "@" not in email or "." not in email:
            raise InvalidEmailError(email)
        self.email:  str = email


# Example usager
try:
    # Trying to create a person with invalid age
    person1 = Person("Alice", -25, "alice@example.com")
except InvalidAgeError as e:
    print(e)
```

```python
# Example usager
try:
    # Trying to create a person with invalid email
    person2 = Person("Bob", 30, "alice#example.com")
except InvalidEmailError as e:
    print(e)
```
